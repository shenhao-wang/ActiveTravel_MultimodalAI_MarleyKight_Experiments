```python
"""
Tampa Active Travel — Generalized Double Machine Learning (DML)
===============================================================
Tests the CAUSAL EFFECT of zoning text embeddings on active travel
mode choice, after controlling for residential self-selection and
sociodemographic confounders.

Three scalar estimands + one vector estimand:
  T1  Residential zone (binary) — cleanest identification
  T2  Pedestrian orientation score (continuous [0,1])
  T3  Front setback (continuous, feet)
  B   32-dim embedding vector (Generalised DML)
  C   PCA on the effect vector → causal direction in embedding space

Changes from prior version:
  - All three scalar treatments in one run (was only T2)
  - Nuisance specs match paper: RF 500 trees, max_depth=12, min_leaf=5
  - Bootstrap SEs (B=500) instead of HC1 only
  - OLS final stage for vector DML (was Ridge α=0.01)
  - Benjamini-Hochberg FDR correction on 32-dim test
  - StratifiedKFold instead of KFold
  - Removed global pre-standardization (was double-dipping)
  - Subgroup analysis for all scalar treatments
  - Safe n_folds floor for small subgroups

References:
  Chernozhukov et al. (2018) Double/debiased ML. Econometrics J.
  Nachtigall et al. (2025) TRD 140, 104593.
"""

import os
import warnings
# WHY: Suppresses sklearn convergence warnings for clean notebook output
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.base import clone
from sklearn.ensemble import (GradientBoostingClassifier,
                               GradientBoostingRegressor,
                               RandomForestClassifier,
                               RandomForestRegressor)
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ─────────────────────────────────────────────────────────────
# CONFIGURATION
# ─────────────────────────────────────────────────────────────
TRIPS_PATH   = 'tampatrips_1.csv'
ZON_CSV_PATH = 'processed_data/tpazoning_clean.csv'
FEAT_PATH    = 'zone_features_combined.csv'
EMB_PATH     = 'zone_embeddings.csv'
ACS_PATH     = 'acs_blockgroup_tampa.csv'

N_FOLDS      = 5
RANDOM_STATE = 16
ALPHA        = 0.05
# WHY B=500: Full bootstrap for this notebook; production scripts use B=50
# after verifying <5% SE difference to reduce SLURM cost.
N_BOOTSTRAP  = 500     # bootstrap replications for SEs

RESIDENTIAL = [
    'RS-150','RS-100','RS-75','RS-60','RS-50',
    'RM-12','RM-16','RM-18','RM-24','RM-35','RM-50','RM-75',
    'RO','RO-1','SH-RS','SH-RS-A','SH-RM','SH-RO','SH-PD',
    'YC-2','YC-4','YC-8','YC-9',
]
NON_RESIDENTIAL = [
    'CG','CI','CN','OP','OP-1','IG','IH','CBD-1','CBD-2',
    'CD-1','CD-2','CD-3','NMU-35','NMU-24','NMU-16',
    'SH-CG','SH-CI','SH-CN','YC-1','YC-3','YC-5','YC-6','YC-7',
]

# ─────────────────────────────────────────────────────────────
# 1. LOAD DATA
# ─────────────────────────────────────────────────────────────
print('=' * 65)
print('  Tampa Active Travel — Generalized DML')
print('  Causal Effects of Zoning on Active Travel')
print('=' * 65)

trips    = pd.read_csv(TRIPS_PATH)
zon_csv  = pd.read_csv(ZON_CSV_PATH).dropna(subset=['ZONECLASS'])
zone_feat = pd.read_csv(FEAT_PATH)
zone_emb = pd.read_csv(EMB_PATH)

trips['active_travel'] = trips['mode_type'].isin([1, 2]).astype(int)
area_col = ('ShapeSTAre_corrected' if 'ShapeSTAre_corrected' in zon_csv.columns
            else 'ShapeSTAre')
print(f'Area column: {area_col}')
print(f'Trips: {len(trips):,}  |  Active travel: {trips["active_travel"].mean():.1%}')

# ─────────────────────────────────────────────────────────────
# 2. BUILD BLOCK-GROUP FEATURES
# ─────────────────────────────────────────────────────────────
print('\nBuilding block-group features...')

dominant = (zon_csv.sort_values(area_col, ascending=False)
            .drop_duplicates(subset='GEOID', keep='first')[['GEOID', 'ZONECLASS']]
            .rename(columns={'ZONECLASS': 'dominant_zone'}))
dominant['zone_context'] = 'Other/PD'
dominant.loc[dominant['dominant_zone'].isin(RESIDENTIAL),     'zone_context'] = 'Residential'
dominant.loc[dominant['dominant_zone'].isin(NON_RESIDENTIAL), 'zone_context'] = 'Non-Residential'

# Binary residential indicator for T1
dominant['residential_zone'] = dominant['dominant_zone'].isin(RESIDENTIAL).astype(int)

def weighted_mean(group, val_col):
    mask = group[val_col].notna()
    if mask.sum() == 0:
        return np.nan
    return np.average(group.loc[mask, val_col], weights=group.loc[mask, area_col])

zon_m = zon_csv.merge(
    zone_feat[['zone_class', 'ped_score_norm', 'front_setback',
               'lot_width', 'max_height']],
    left_on='ZONECLASS', right_on='zone_class', how='left')

def entropy(group):
    total = group[area_col].sum()
    if total == 0:
        return 0.0
    p = group[area_col] / total
    p = p[p > 0]
    return float(-np.sum(p * np.log(p)))

scraper_agg = (zon_m.groupby('GEOID')
               .apply(lambda g: pd.Series({
                   'ped_score_weighted': weighted_mean(g, 'ped_score_norm'),
                   'front_setback_avg':  weighted_mean(g, 'front_setback'),
                   'lot_width_avg':      weighted_mean(g, 'lot_width'),
                   'max_height_avg':     weighted_mean(g, 'max_height'),
                   'zoning_entropy':     entropy(g),
               }))
               .reset_index())

emb_cols = [c for c in zone_emb.columns if c.startswith('emb_')]
emb_lookup = zone_emb.set_index('zone_class')[emb_cols].to_dict('index')
bg_emb_rows = []
for geoid, grp in zon_csv.groupby('GEOID'):
    total = grp[area_col].sum()
    vec, cov = np.zeros(len(emb_cols)), 0.0
    for _, row in grp.iterrows():
        z = row['ZONECLASS']
        if z in emb_lookup:
            w    = row[area_col] / total
            vec += np.array(list(emb_lookup[z].values())) * w
            cov += w
    if cov > 0:
        bg_emb_rows.append({'GEOID': geoid, **dict(zip(emb_cols, vec))})
bg_emb = pd.DataFrame(bg_emb_rows)

bg_features = (dominant
               .merge(scraper_agg, on='GEOID', how='left')
               .merge(bg_emb,      on='GEOID', how='left'))

# ACS (optional)
acs_cols = []
if os.path.exists(ACS_PATH):
    acs = pd.read_csv(ACS_PATH)
    ACS_VARS = ['median_hh_income', 'pct_bachelors_plus',
                'pct_white_nonhisp', 'pct_black_nonhisp', 'pct_hispanic',
                'pct_poverty', 'pct_owner_occ']
    acs_cols = [c for c in ACS_VARS if c in acs.columns]
    bg_features = bg_features.merge(acs[['GEOID'] + acs_cols],
                                     on='GEOID', how='left')
    print(f'ACS loaded: {len(acs_cols)} neighbourhood variables added')
else:
    print(f'ACS file not found at {ACS_PATH} — skipping neighbourhood confounders')

df = trips.merge(bg_features, left_on='o_bg', right_on='GEOID', how='left')
df = df.dropna(subset=['zoning_entropy']).reset_index(drop=True)
print(f'Analysis sample: {len(df):,} trips  |  AT rate: {df["active_travel"].mean():.1%}')

# ─────────────────────────────────────────────────────────────
# 3. FEATURE SETS & TREATMENTS
# ─────────────────────────────────────────────────────────────
INDIVIDUAL_CONFOUNDERS = [
    'age', 'gender', 'education', 'income_detailed',
    'num_veh', 'hhsize', 'trip_distance', 'trip_duration',
    'd_purpose_category_imputed',
]
W_COLS    = INDIVIDUAL_CONFOUNDERS + acs_cols
Y_COL     = 'active_travel'

# All three scalar treatments
SCALAR_TREATMENTS = {
    'T1: Residential zone': {
        'col': 'residential_zone',
        'type': 'binary',
        'description': 'Binary indicator: 1=residential dominant zone, 0=non-residential',
    },
    'T2: Ped orientation score': {
        'col': 'ped_score_weighted',
        'type': 'continuous',
        'description': 'Area-weighted pedestrian rubric [0,1]',
    },
    'T3: Front setback': {
        'col': 'front_setback_avg',
        'type': 'continuous',
        'description': 'Area-weighted front setback (feet)',
    },
}

# ─────────────────────────────────────────────────────────────
# 4. DML CORE FUNCTIONS
# ─────────────────────────────────────────────────────────────

def impute_scale(X_tr, X_te):
    """Impute (median) and standardise. Fit on train, apply to test.
    WHY per-fold: Per-fold standardization to avoid information leakage
    across folds."""
    imp = SimpleImputer(strategy='median')
    sc  = StandardScaler()
    X_tr_out = sc.fit_transform(imp.fit_transform(X_tr))
    X_te_out = sc.transform(imp.transform(X_te))
    return X_tr_out, X_te_out


def cross_fit_residuals(Y, T, W, clf_Y, reg_T, n_folds=N_FOLDS,
                         seed=RANDOM_STATE, stratify_on=None):
    """
    Cross-fitted DML residuals for scalar treatment T.
    Uses StratifiedKFold on the outcome for balanced folds.
    WHY: Stratify on AT outcome (~10% prevalence) to ensure each fold
    has balanced class representation.
    """
    n     = len(Y)
    Y_res = np.zeros(n)
    T_res = np.zeros(n)

    if stratify_on is not None:
        kf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=seed)
        split_iter = kf.split(W, stratify_on)
    else:
        kf = KFold(n_splits=n_folds, shuffle=True, random_state=seed)
        split_iter = kf.split(W)

    for fold, (tr_idx, te_idx) in enumerate(split_iter):
        W_tr, W_te = impute_scale(W[tr_idx], W[te_idx])

        # Impute and scale T per fold (no global pre-standardization)
        imp_T = SimpleImputer(strategy='median')
        sc_T  = StandardScaler()
        T_tr  = sc_T.fit_transform(imp_T.fit_transform(
                    T[tr_idx].reshape(-1, 1))).ravel()
        T_te  = sc_T.transform(imp_T.transform(
                    T[te_idx].reshape(-1, 1))).ravel()

        clf_Y_f = clone(clf_Y)
        reg_T_f = clone(reg_T)
        clf_Y_f.fit(W_tr, Y[tr_idx])
        reg_T_f.fit(W_tr, T_tr)

        Y_hat = clf_Y_f.predict_proba(W_te)[:, 1]
        T_hat = reg_T_f.predict(W_te)
        Y_res[te_idx] = Y[te_idx] - Y_hat
        T_res[te_idx] = T_te - T_hat

    return Y_res, T_res


def dml_estimate(Y_res, T_res):
    """
    Final DML stage: OLS of Y_res on T_res (no intercept).
    Returns theta (point estimate).
    """
    T_mat = T_res.reshape(-1, 1)
    reg   = LinearRegression(fit_intercept=False)
    reg.fit(T_mat, Y_res)
    return reg.coef_[0]


def dml_hc1_se(Y_res, T_res, theta):
    """HC1 sandwich standard error. Returns inf if T_res has no variance."""
    n   = len(Y_res)
    eps = Y_res - theta * T_res
    T2  = (T_res ** 2).sum()
    if T2 < 1e-12:
        return np.inf
    return np.sqrt(n / (n - 1) * np.sum((T_res * eps) ** 2) / T2 ** 2)


def dml_scalar_with_bootstrap(Y, T, W, clf_Y, reg_T, n_folds=N_FOLDS,
                                n_boot=N_BOOTSTRAP, seed=RANDOM_STATE,
                                stratify_on=None):
    """
    Full scalar DML with bootstrap standard errors.

    1. Point estimate from cross-fitted residuals on full sample.
    2. Bootstrap SE: resample observations B times, re-run full
       cross-fitting on each resample.

    Returns dict with theta, se_hc1, se_boot, p, ci_lo, ci_hi.
    """
    # Point estimate
    Y_res, T_res = cross_fit_residuals(Y, T, W, clf_Y, reg_T,
                                        n_folds=n_folds, seed=seed,
                                        stratify_on=stratify_on)
    theta = dml_estimate(Y_res, T_res)
    se_hc1 = dml_hc1_se(Y_res, T_res, theta)

    # Bootstrap
    n = len(Y)
    rng = np.random.RandomState(seed)
    boot_thetas = np.zeros(n_boot)

    for b in range(n_boot):
        if (b + 1) % 100 == 0:
            print(f'    Bootstrap {b+1}/{n_boot}...')
        idx_b = rng.choice(n, size=n, replace=True)
        Y_b, T_b, W_b = Y[idx_b], T[idx_b], W[idx_b]
        strat_b = stratify_on[idx_b] if stratify_on is not None else None
        try:
            Y_r_b, T_r_b = cross_fit_residuals(
                Y_b, T_b, W_b, clf_Y, reg_T,
                n_folds=n_folds, seed=seed + b,
                stratify_on=strat_b)
            boot_thetas[b] = dml_estimate(Y_r_b, T_r_b)
        except Exception:
            boot_thetas[b] = np.nan

    boot_thetas = boot_thetas[~np.isnan(boot_thetas)]
    se_boot = np.std(boot_thetas, ddof=1) if len(boot_thetas) > 1 else np.nan

    # Use bootstrap SE for inference; fall back to HC1 if bootstrap failed
    se = se_boot if (np.isfinite(se_boot) and se_boot > 1e-12) else se_hc1
    z_crit  = stats.norm.ppf(1 - ALPHA / 2)
    t_stat  = theta / se if se > 1e-12 else np.nan
    p_value = 2 * stats.norm.sf(np.abs(t_stat)) if np.isfinite(t_stat) else np.nan
    ci_lo   = theta - z_crit * se
    ci_hi   = theta + z_crit * se

    return {
        'theta': theta, 'se_hc1': se_hc1, 'se_boot': se_boot,
        't_stat': t_stat, 'p_value': p_value,
        'ci_lo': ci_lo, 'ci_hi': ci_hi,
        'n_boot_valid': len(boot_thetas),
    }


def cross_fit_vector_residuals(Y, T_matrix, W, clf_Y, reg_T_proto,
                                n_folds=N_FOLDS, seed=RANDOM_STATE,
                                stratify_on=None):
    """
    Cross-fitted residuals for VECTOR treatment (n × d).
    Each dimension of T is partialled out independently.
    """
    n, d  = T_matrix.shape
    Y_res = np.zeros(n)
    T_res = np.zeros((n, d))

    if stratify_on is not None:
        kf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=seed)
        split_iter = kf.split(W, stratify_on)
    else:
        kf = KFold(n_splits=n_folds, shuffle=True, random_state=seed)
        split_iter = kf.split(W)

    for fold, (tr_idx, te_idx) in enumerate(split_iter):
        W_tr, W_te = impute_scale(W[tr_idx], W[te_idx])

        # Impute and scale T per fold
        imp_T = SimpleImputer(strategy='median')
        sc_T  = StandardScaler()
        T_tr  = sc_T.fit_transform(imp_T.fit_transform(T_matrix[tr_idx]))
        T_te  = sc_T.transform(imp_T.transform(T_matrix[te_idx]))

        # Outcome nuisance
        clf_f = clone(clf_Y)
        clf_f.fit(W_tr, Y[tr_idx])
        Y_hat = clf_f.predict_proba(W_te)[:, 1]
        Y_res[te_idx] = Y[te_idx] - Y_hat

        # Treatment nuisance: one regressor per dimension
        for dim in range(d):
            reg_f = clone(reg_T_proto)
            reg_f.fit(W_tr, T_tr[:, dim])
            T_hat_dim = reg_f.predict(W_te)
            T_res[te_idx, dim] = T_te[:, dim] - T_hat_dim

        print(f'  Fold {fold+1}/{n_folds} complete')

    return Y_res, T_res


def dml_vector_final_stage(Y_res, T_res):
    """
    Final stage for vector DML: OLS regression of Y_res on T_res.
    PCA dimensions are orthogonal → OLS is well-conditioned.
    Sandwich SEs per dimension.
    Applies Benjamini-Hochberg FDR correction.
    """
    n, d   = T_res.shape

    # WHY OLS instead of Ridge: Dims are orthogonal by PCA construction,
    # so OLS is well-conditioned. No intercept because both residuals
    # are mean-zero by construction.
    reg = LinearRegression(fit_intercept=False)
    reg.fit(T_res, Y_res)
    thetas = reg.coef_

    eps    = Y_res - T_res @ thetas
    z_crit = stats.norm.ppf(1 - ALPHA / 2)

    rows = []
    raw_pvals = []
    for j in range(d):
        t_j   = T_res[:, j]
        T2_j  = (t_j ** 2).sum()
        se_j  = np.sqrt(n / (n - 1) * np.sum((t_j * eps) ** 2) / T2_j ** 2)
        t_stat = thetas[j] / se_j
        p_val  = 2 * stats.t.sf(np.abs(t_stat), df=n - 1)
        raw_pvals.append(p_val)
        rows.append({
            'dim':      f'emb_{j:03d}',
            'coef':     thetas[j],
            'se':       se_j,
            't_stat':   t_stat,
            'p_raw':    p_val,
            'ci_lo':    thetas[j] - z_crit * se_j,
            'ci_hi':    thetas[j] + z_crit * se_j,
        })

    result_df = pd.DataFrame(rows)

    # WHY Benjamini-Hochberg: Controls false discovery rate across 32
    # embedding dimensions within this single specification.
    try:
        from statsmodels.stats.multitest import multipletests
        reject, p_adj, _, _ = multipletests(raw_pvals, alpha=ALPHA, method='fdr_bh')
    except ImportError:
        # Manual BH if statsmodels unavailable
        m = len(raw_pvals)
        sorted_idx = np.argsort(raw_pvals)
        p_adj = np.zeros(m)
        for rank, idx in enumerate(sorted_idx):
            p_adj[idx] = raw_pvals[idx] * m / (rank + 1)
        # Enforce monotonicity (step-up)
        for i in range(m - 2, -1, -1):
            idx = sorted_idx[i]
            idx_next = sorted_idx[i + 1]
            p_adj[idx] = min(p_adj[idx], p_adj[idx_next])
        p_adj = np.minimum(p_adj, 1.0)
        reject = p_adj < ALPHA
    result_df['p_bh']   = p_adj
    result_df['sig_bh'] = reject
    # Also keep raw significance for comparison
    result_df['sig_raw'] = result_df['p_raw'] < ALPHA

    return result_df


# ─────────────────────────────────────────────────────────────
# 5. PREPARE ARRAYS (no global standardization — done per fold)
# ─────────────────────────────────────────────────────────────
print('\nPreparing arrays...')

avail_W = [c for c in W_COLS if c in df.columns]
W_raw   = df[avail_W].values.astype(float)
Y       = df[Y_COL].values.astype(float)
T_vec   = df[emb_cols].values.astype(float)

print(f'W features ({len(avail_W)}): {avail_W}')
print(f'Y shape: {Y.shape}  |  T_vector shape: {T_vec.shape}')

# WHY RF over lasso for nuisance: Nonlinear confounding favors trees.
# Nuisance models — matched to paper specs
clf_Y = RandomForestClassifier(
    n_estimators=500, max_depth=12, min_samples_leaf=5,
    random_state=RANDOM_STATE, n_jobs=-1)
reg_T = RandomForestRegressor(
    n_estimators=500, max_depth=12, min_samples_leaf=5,
    random_state=RANDOM_STATE, n_jobs=-1)

# Robustness nuisance (GB)
clf_Y_gb = GradientBoostingClassifier(
    n_estimators=500, learning_rate=0.05, max_depth=3,
    random_state=RANDOM_STATE)
reg_T_gb = GradientBoostingRegressor(
    n_estimators=500, learning_rate=0.05, max_depth=3,
    random_state=RANDOM_STATE)


# ─────────────────────────────────────────────────────────────
# 6. ALL SCALAR TREATMENTS (T1, T2, T3)
# ─────────────────────────────────────────────────────────────
scalar_results = []

for t_label, t_info in SCALAR_TREATMENTS.items():
    t_col = t_info['col']
    if t_col not in df.columns:
        print(f'\n  WARNING: {t_col} not found in df — skipping {t_label}')
        continue

    T_sc = df[t_col].values.astype(float)

    print(f'\n\n{"=" * 65}')
    print(f'  {t_label}')
    print(f'  Treatment: {t_col}  ({t_info["type"]})')
    print(f'  {t_info["description"]}')
    print(f'  Nuisance: RF (500 trees, depth 12, {N_FOLDS}-fold cross-fitting)')
    print(f'  Bootstrap: B={N_BOOTSTRAP}')
    print(f'{"=" * 65}')

    # ── Primary estimate (RF nuisance + bootstrap SE) ──
    print(f'\n  Primary estimate (RF nuisance):')
    result_rf = dml_scalar_with_bootstrap(
        Y, T_sc, W_raw, clf_Y, reg_T,
        n_folds=N_FOLDS, n_boot=N_BOOTSTRAP,
        seed=RANDOM_STATE, stratify_on=Y.astype(int))

    print(f'\n  ── {t_label} Results ──')
    print(f'  θ  = {result_rf["theta"]:+.4f}')
    print(f'  SE (HC1)      = {result_rf["se_hc1"]:.4f}')
    print(f'  SE (bootstrap)= {result_rf["se_boot"]:.4f}  (B={result_rf["n_boot_valid"]})')
    print(f'  t-statistic   = {result_rf["t_stat"]:+.3f}')
    print(f'  p-value       = {result_rf["p_value"]:.4f}')
    print(f'  95% CI        = [{result_rf["ci_lo"]:+.4f}, {result_rf["ci_hi"]:+.4f}]')
    sig_str = 'YES ✓' if result_rf['p_value'] < ALPHA else 'NO'
    print(f'  Significant:    {sig_str}')

    scalar_results.append({
        'treatment': t_label, 'col': t_col, 'nuisance': 'RF',
        'sample': 'Full', 'n': len(Y),
        **{k: result_rf[k] for k in ['theta','se_hc1','se_boot',
                                       't_stat','p_value','ci_lo','ci_hi']},
    })

    # ── Robustness check (GB nuisance) ──
    print(f'\n  Robustness check (GB nuisance):')
    result_gb = dml_scalar_with_bootstrap(
        Y, T_sc, W_raw, clf_Y_gb, reg_T_gb,
        n_folds=N_FOLDS, n_boot=min(100, N_BOOTSTRAP),  # fewer reps for GB
        seed=RANDOM_STATE, stratify_on=Y.astype(int))

    same_sig = (result_rf['p_value'] < ALPHA) == (result_gb['p_value'] < ALPHA)
    rob_str = 'ROBUST ✓' if same_sig else 'SENSITIVITY ⚠'
    print(f'  θ (GB) = {result_gb["theta"]:+.4f}  SE={result_gb["se_boot"]:.4f}  '
          f'p={result_gb["p_value"]:.4f}  {rob_str}')

    scalar_results.append({
        'treatment': t_label, 'col': t_col, 'nuisance': 'GB',
        'sample': 'Full', 'n': len(Y),
        **{k: result_gb[k] for k in ['theta','se_hc1','se_boot',
                                       't_stat','p_value','ci_lo','ci_hi']},
    })

    # ── Subgroup analysis (residential / non-residential) ──
    # Skip for binary context treatments (T1): the treatment collapses
    # to a near-constant within each subgroup, making DML undefined.
    if t_info['type'] == 'binary':
        print(f'\n  Subgroup analysis: skipped for binary treatment '
              f'(T ≈ constant within context subgroups)')
        continue

    print(f'\n  Subgroup analysis:')
    for sub_label, mask_val in [('Residential', 'Residential'),
                                 ('Non-Residential', 'Non-Residential')]:
        mask = (df['zone_context'] == mask_val).values
        n_sub = mask.sum()
        if n_sub < 200:
            print(f'    {sub_label}: too few obs ({n_sub}), skipping')
            continue

        Y_sub, T_sub, W_sub = Y[mask], T_sc[mask], W_raw[mask]
        # Safe fold count: at least 2, at most N_FOLDS
        n_folds_sub = max(2, min(N_FOLDS, n_sub // 100))

        # Subgroups use HC1 only (bootstrap on subgroups is very slow)
        Y_r, T_r = cross_fit_residuals(
            Y_sub, T_sub, W_sub, clf_Y, reg_T,
            n_folds=n_folds_sub, seed=RANDOM_STATE,
            stratify_on=Y_sub.astype(int))
        th = dml_estimate(Y_r, T_r)
        se = dml_hc1_se(Y_r, T_r, th)
        t_s = th / se
        p_s = 2 * stats.norm.sf(np.abs(t_s))
        z_c = stats.norm.ppf(1 - ALPHA / 2)

        sig_s = '*' if p_s < ALPHA else ''
        print(f'    {sub_label} (n={n_sub:,}):  '
              f'θ={th:+.4f}  SE(HC1)={se:.4f}  p={p_s:.4f} {sig_s}')

        scalar_results.append({
            'treatment': t_label, 'col': t_col, 'nuisance': 'RF',
            'sample': sub_label, 'n': n_sub,
            'theta': th, 'se_hc1': se, 'se_boot': np.nan,
            't_stat': t_s, 'p_value': p_s,
            'ci_lo': th - z_c * se, 'ci_hi': th + z_c * se,
        })

# Save scalar results
scalar_df = pd.DataFrame(scalar_results)
scalar_df.to_csv('dml_scalar_results.csv', index=False)
print('\n\nSaved: dml_scalar_results.csv')

# ── OLS comparison for scalar treatments ──
# NOTE: OLS standardises T globally; DML standardises per-fold.
# Per-fold SDs ≈ global SD at n=5749, so the bias estimate is
# approximate but close. Both are in per-SD-of-T units.
print('\n── OLS Comparison (for bias quantification) ──')
print(f'  {"Treatment":<28} {"OLS θ":>9} {"DML θ":>9} {"Bias":>9} {"Bias %":>9}')
print('  ' + '─' * 70)

for t_label, t_info in SCALAR_TREATMENTS.items():
    t_col = t_info['col']
    if t_col not in df.columns:
        continue
    T_sc = df[t_col].values.astype(float)
    # Simple OLS with full W
    imp_ols = SimpleImputer(strategy='median')
    sc_ols  = StandardScaler()
    W_ols   = sc_ols.fit_transform(imp_ols.fit_transform(W_raw))
    T_ols   = StandardScaler().fit_transform(
                  SimpleImputer(strategy='median').fit_transform(
                      T_sc.reshape(-1, 1))).ravel()
    X_ols   = np.column_stack([T_ols, W_ols])
    reg_ols = LinearRegression()
    reg_ols.fit(X_ols, Y)
    ols_theta = reg_ols.coef_[0]

    dml_row = scalar_df[(scalar_df['treatment'] == t_label) &
                         (scalar_df['nuisance'] == 'RF') &
                         (scalar_df['sample'] == 'Full')]
    if len(dml_row) > 0:
        dml_theta = dml_row.iloc[0]['theta']
        bias = ols_theta - dml_theta
        bias_pct = (bias / abs(dml_theta) * 100) if dml_theta != 0 else np.nan
        print(f'  {t_label:<28} {ols_theta:>+9.4f} {dml_theta:>+9.4f} '
              f'{bias:>+9.4f} {bias_pct:>+8.1f}%')

# ── Scalar figure ──
fig_a, ax_a = plt.subplots(figsize=(10, 5))
full_rf = scalar_df[(scalar_df['nuisance'] == 'RF') &
                     (scalar_df['sample'] == 'Full')]
labels = full_rf['treatment'].tolist()
thetas_plot = full_rf['theta'].values
ci_lo_plot  = full_rf['ci_lo'].values
ci_hi_plot  = full_rf['ci_hi'].values
colors_a = ['#4878CF', '#6ACC65', '#D65F5F']

for i, (lab, th, lo, hi, col) in enumerate(
        zip(labels, thetas_plot, ci_lo_plot, ci_hi_plot, colors_a)):
    ax_a.barh(i, th, xerr=[[th - lo], [hi - th]], color=col,
              alpha=0.85, height=0.5, capsize=6,
              error_kw={'linewidth': 2, 'capthick': 2})
ax_a.axvline(0, color='black', linewidth=0.8, linestyle='--')
ax_a.set_yticks(range(len(labels)))
ax_a.set_yticklabels(labels)
ax_a.set_xlabel('DML Coefficient θ (with 95% bootstrap CI)')
ax_a.set_title('Scalar DML: Causal Effects of Zoning Features on P(Active Travel)\n'
               f'RF nuisance (500 trees, depth 12)  ·  Bootstrap B={N_BOOTSTRAP}',
               fontweight='bold')
ax_a.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('fig_dml_scalar.png', dpi=150, bbox_inches='tight')
plt.close()
print('\nSaved: fig_dml_scalar.png')


# ─────────────────────────────────────────────────────────────
# 7. ESTIMAND B — VECTOR DML (embedding → active travel)
# ─────────────────────────────────────────────────────────────
print('\n\n' + '=' * 65)
print('  ESTIMAND B: Generalized DML — Embedding Vector Treatment')
print('=' * 65)
print(f'  Treatment: {len(emb_cols)}-dimensional zoning text embedding')
print(f'  Nuisance: RF (500 trees, depth 12, {N_FOLDS}-fold cross-fitting)')
print(f'  FDR correction: Benjamini-Hochberg at α={ALPHA}')
print()

Y_res_B, T_res_B = cross_fit_vector_residuals(
    Y, T_vec, W_raw, clf_Y, reg_T,
    n_folds=N_FOLDS, seed=RANDOM_STATE,
    stratify_on=Y.astype(int))
results_B = dml_vector_final_stage(Y_res_B, T_res_B)
results_B.to_csv('dml_results.csv', index=False)
print('\nSaved: dml_results.csv')

n_sig_raw = results_B['sig_raw'].sum()
n_sig_bh  = results_B['sig_bh'].sum()
print(f'\n  ── Vector DML Results ──')
print(f'  Significant dims (raw p < {ALPHA}):    {n_sig_raw} / {len(results_B)}')
print(f'  Significant dims (BH-corrected):  {n_sig_bh} / {len(results_B)}')
print(f'\n  Top 10 by |coefficient|:')
print(f'  {"Dim":<10} {"Coef":>8} {"SE":>8} {"t":>7} {"p(raw)":>8} '
      f'{"p(BH)":>8} {"Sig":>5}')
print(f'  {"─"*60}')
top10 = results_B.reindex(
    results_B['coef'].abs().sort_values(ascending=False).index).head(10)
for _, row in top10.iterrows():
    sig_mark = '**' if row['sig_bh'] else ('*' if row['sig_raw'] else ' ')
    print(f'  {row["dim"]:<10} {row["coef"]:>+8.4f} {row["se"]:>8.4f} '
          f'{row["t_stat"]:>+7.2f} {row["p_raw"]:>8.4f} '
          f'{row["p_bh"]:>8.4f} {sig_mark:>5}')

# Figure B
fig_b, ax_b = plt.subplots(figsize=(14, 5))
dims_plot  = results_B['dim'].tolist()
coefs_plot = results_B['coef'].values
ses_plot   = results_B['se'].values
xs = np.arange(len(dims_plot))
colors_b = ['#D65F5F' if results_B.iloc[i]['sig_bh'] else
            ('#FFB347' if results_B.iloc[i]['sig_raw'] else '#AAAAAA')
            for i in range(len(results_B))]

ax_b.bar(xs, coefs_plot, color=colors_b, alpha=0.85, width=0.7)
ax_b.errorbar(xs, coefs_plot, yerr=1.96 * ses_plot, fmt='none',
              color='black', linewidth=0.8, capsize=3)
ax_b.axhline(0, color='black', linewidth=0.8, linestyle='--')
ax_b.set_xticks(xs[::4])
ax_b.set_xticklabels([dims_plot[i] for i in range(0, len(dims_plot), 4)],
                      rotation=45, ha='right', fontsize=8)
ax_b.set_xlabel('Embedding Dimension')
ax_b.set_ylabel('DML Coefficient θ')
ax_b.set_title('Generalized DML: Causal Effects of Zoning Text Embedding Dimensions\n'
               f'Red = BH-significant · Orange = raw-significant only · '
               f'Grey = n.s. · 95% CI',
               fontweight='bold')
ax_b.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('fig_dml_vector.png', dpi=150, bbox_inches='tight')
plt.close()
print('Saved: fig_dml_vector.png')


# ─────────────────────────────────────────────────────────────
# 8. ESTIMAND C — PCA ON EFFECT VECTOR
# ─────────────────────────────────────────────────────────────
print('\n' + '=' * 65)
print('  ESTIMAND C: PCA on Effect Vector')
print('  Which regulatory direction drives active travel?')
print('=' * 65)

emb_matrix  = zone_emb[emb_cols].values
zone_names  = zone_emb['zone_class'].tolist()
coef_vector = results_B['coef'].values
# WHY zone causal scores: Project each zone0027s embedding onto the DML
# coefficient vector to get a single scalar measuring AT-promotion.
causal_scores = emb_matrix @ coef_vector

zone_causal = pd.DataFrame({
    'zone_class':   zone_names,
    'causal_score': causal_scores,
}).sort_values('causal_score', ascending=False)

print('\n  Zones with highest positive causal score:')
for _, row in zone_causal.head(8).iterrows():
    print(f'    {row["zone_class"]:<12}  score={row["causal_score"]:+.4f}')

print('\n  Zones with most negative causal score:')
for _, row in zone_causal.tail(8).iterrows():
    print(f'    {row["zone_class"]:<12}  score={row["causal_score"]:+.4f}')

zone_causal.to_csv('dml_zone_causal_scores.csv', index=False)
print('\nSaved: dml_zone_causal_scores.csv')

# 2D PCA coloured by causal score
pca2   = PCA(n_components=2, random_state=RANDOM_STATE)
coords = pca2.fit_transform(emb_matrix)

fig_c, ax_c = plt.subplots(figsize=(11, 8))
ax_c.set_facecolor('#FAFAFA')
ax_c.grid(alpha=0.3, linewidth=0.5)
sc_plot = ax_c.scatter(coords[:, 0], coords[:, 1],
                        c=causal_scores, cmap='RdBu_r',
                        s=100, alpha=0.9, zorder=3,
                        edgecolors='white', linewidths=0.5,
                        vmin=-np.abs(causal_scores).max(),
                        vmax= np.abs(causal_scores).max())
plt.colorbar(sc_plot, ax=ax_c,
             label='DML causal score\n(+) → more AT  (–) → more car travel',
             shrink=0.7)
for i, zone in enumerate(zone_names):
    ax_c.annotate(zone, (coords[i, 0], coords[i, 1]),
                  fontsize=5.5, ha='center', va='bottom',
                  xytext=(0, 4), textcoords='offset points', alpha=0.9)
ax_c.set_xlabel(f'PC1 ({pca2.explained_variance_ratio_[0]:.1%} variance)')
ax_c.set_ylabel(f'PC2 ({pca2.explained_variance_ratio_[1]:.1%} variance)')
ax_c.set_title('Causal Direction in Zoning Embedding Space\n'
               'Colour = DML causal effect score on P(Active Travel)',
               fontweight='bold')
plt.tight_layout()
plt.savefig('fig_dml_pca_direction.png', dpi=150, bbox_inches='tight')
plt.close()
print('Saved: fig_dml_pca_direction.png')


# ─────────────────────────────────────────────────────────────
# 9. SUMMARY
# ─────────────────────────────────────────────────────────────
print('\n' + '=' * 65)
print('  DML SUMMARY')
print('=' * 65)

print('\n  ── Scalar Treatments ──')
print(f'  {"Treatment":<28} {"θ":>9} {"SE(boot)":>9} {"p":>8} {"95% CI":>22}')
print('  ' + '─' * 80)
for _, row in scalar_df[(scalar_df['nuisance'] == 'RF') &
                         (scalar_df['sample'] == 'Full')].iterrows():
    se_col = row['se_boot'] if not np.isnan(row['se_boot']) else row['se_hc1']
    sig = '***' if row['p_value'] < 0.001 else (
          '**'  if row['p_value'] < 0.01  else (
          '*'   if row['p_value'] < 0.05  else 'n.s.'))
    print(f'  {row["treatment"]:<28} {row["theta"]:>+9.4f} {se_col:>9.4f} '
          f'{row["p_value"]:>8.4f} [{row["ci_lo"]:>+.4f}, {row["ci_hi"]:>+.4f}] {sig}')

print(f'\n  ── Vector DML ({len(results_B)} dimensions) ──')
print(f'  Significant (BH-corrected): {n_sig_bh} / {len(results_B)}')
sig_dims = results_B[results_B['sig_bh']].sort_values('coef', ascending=False)
if len(sig_dims) > 0:
    print(f'  Most positive (→ AT):  {sig_dims.iloc[0]["dim"]} '
          f'(θ={sig_dims.iloc[0]["coef"]:+.4f})')
    print(f'  Most negative (→ car): {sig_dims.iloc[-1]["dim"]} '
          f'(θ={sig_dims.iloc[-1]["coef"]:+.4f})')

print(f'\n  ── Causal PCA Direction ──')
print(f'  Most AT-promoting zone:  {zone_causal.iloc[0]["zone_class"]} '
      f'(score={zone_causal.iloc[0]["causal_score"]:+.4f})')
print(f'  Most car-promoting zone: {zone_causal.iloc[-1]["zone_class"]} '
      f'(score={zone_causal.iloc[-1]["causal_score"]:+.4f})')

print(f'\n  Outputs:')
print(f'    dml_scalar_results.csv    — all scalar treatment estimates + subgroups')
print(f'    dml_results.csv           — vector DML per-dimension estimates')
print(f'    dml_zone_causal_scores.csv— zone-level causal PCA scores')
print(f'    fig_dml_scalar.png        — scalar treatment effects + CI')
print(f'    fig_dml_vector.png        — embedding dimension coefficients + CI')
print(f'    fig_dml_pca_direction.png — causal direction in embedding space')
print('\n✓ DML analysis complete.')
```
