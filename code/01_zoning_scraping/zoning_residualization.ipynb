```python
"""
Tampa Active Travel ├бРѓгРђЮ Zoning-ACS Residualization Diagnostic
============================================================
Addresses the question: how much of each zoning feature's variance
is explained by neighbourhood demographics (ACS), and does the
zoning-active travel association survive after removing that
demographic signal?

Four outputs:

  1. DIAGNOSTIC TABLE
     For each zoning feature:
       - OLS R├ѓ┬▓: demographic predictability under linear assumption
       - RF R├ѓ┬▓:  demographic predictability under nonlinear assumption
       - OLS-RF gap: how much nonlinearity is in the confounding structure
       - DML ATE (from dml_results.csv if present): attenuated causal estimate
       - Overlap flag: warning if RF R├ѓ┬▓ > 0.7 (positivity concern)

     The OLS R├ѓ┬▓ is reported to show the linear baseline. The RF R├ѓ┬▓ is
     what DML's nuisance model actually partials out, and is the correct
     comparison to the DML ATE. The gap between OLS and RF tells readers
     how much of the confounding structure is nonlinear.

  2. RESIDUALIZED CLASSIFIER COMPARISON
     Reruns the v8/v9 model B (scraper features) and model C (embeddings)
     using RF-residualized zoning features instead of raw features.
     Accuracy and log loss compared to raw-feature models.
     The gap quantifies how much the zoning-AT association was mediated
     by demographic sorting vs. independent of it.

  3. OVERLAP / POSITIVITY DIAGNOSTICS
     For features with RF R├ѓ┬▓ > 0.7, plot treatment distributions across
     ACS strata and compute the effective support of the treatment.
     Unstable DML estimates are most likely where propensity is near 0 or 1.

  4. INTERPRETATION NOTE (printed to stdout)
     Reminds analyst that the residualization and the heterogeneity story
     are not mutually exclusive ├бРѓгРђЮ both predict null full-sample / significant
     residential-subgroup patterns for ped score.

Note on functional form:
  OLS residualization with nonlinear confounding UNDERstates demographic
  entanglement ├бРѓгРђЮ the opposite of what a "look how confounded this is"
  diagnostic should do. RF residualization matches the DML nuisance
  functional form and avoids this problem. Both are reported here so
  the OLS-RF gap is visible.

Note on framing:
  R├ѓ┬▓ here is "demographic predictability of the zoning feature"
  (treatment-confounder entanglement), NOT a summary of DML.
  The DML estimand depends on both treatment-side and outcome-side
  nuisance; R├ѓ┬▓ only speaks to the treatment side.

Required files:
  tampatrips_1.csv
  tpazoning_clean.csv (ShapeSTAre_corrected)
  zone_features_combined.csv
  zone_embeddings.csv
  acs_blockgroup_tampa.csv
  dml_results.csv              (optional, for ATE column)

Outputs:
  zoning_acs_diagnostic_table.csv
  fig_zoning_acs_r2.png
  fig_overlap_<feature>.png    (one per high-R├ѓ┬▓ feature)
  fig_residual_classifier_comparison.png
  residual_classifier_results.csv
"""

import warnings
warnings.filterwarnings('ignore')

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from sklearn.base import clone
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor, GradientBoostingClassifier
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import (accuracy_score, log_loss, r2_score,
                             roc_auc_score, average_precision_score,
                             brier_score_loss)
from sklearn.calibration import calibration_curve
from sklearn.model_selection import KFold, train_test_split, GroupShuffleSplit
from sklearn.preprocessing import StandardScaler

# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# CONFIGURATION
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
TRIPS_PATH   = 'tampatrips_1.csv'
ZON_CSV_PATH = 'processed_data/tpazoning_clean.csv'
FEAT_PATH    = 'zone_features_combined.csv'
EMB_PATH     = 'zone_embeddings.csv'
ACS_PATH     = 'acs_blockgroup_tampa.csv'
DML_PATH     = 'dml_results.csv'           # optional

RANDOM_STATE = 16
N_FOLDS      = 5
# 0.70 threshold: features with RF R├ѓ┬▓ above this are so demographically
# predicted that DML positivity (treatment variation within strata) is at risk.
OVERLAP_THRESHOLD = 0.70   # RF R├ѓ┬▓ above this ├бРђаРђЎ flag for positivity check

ACS_FEATURES = [
    'median_hh_income', 'pct_bachelors_plus',
    'pct_white_nonhisp', 'pct_black_nonhisp', 'pct_hispanic',
    'pct_poverty', 'pct_owner_occ',
]

# Zoning features to diagnose
ZONING_FEATURES = [
    'ped_score_weighted',
    'front_setback_avg',
    'lot_width_avg',
    'max_height_avg',
    'zoning_entropy',
]

# Zone context
RESIDENTIAL = [
    'RS-150','RS-100','RS-75','RS-60','RS-50',
    'RM-12','RM-16','RM-18','RM-24','RM-35','RM-50','RM-75',
    'RO','RO-1','SH-RS','SH-RS-A','SH-RM','SH-RO','SH-PD',
    'YC-2','YC-4','YC-8','YC-9',
]
NON_RESIDENTIAL = [
    'CG','CI','CN','OP','OP-1','IG','IH','CBD-1','CBD-2',
    'CD-1','CD-2','CD-3','NMU-35','NMU-24','NMU-16',
    'SH-CG','SH-CI','SH-CN','YC-1','YC-3','YC-5','YC-6','YC-7',
]

# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# 1. LOAD DATA
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
print('=' * 65)
print('  Zoning-ACS Residualization Diagnostic')
print('=' * 65)

trips    = pd.read_csv(TRIPS_PATH)
zon_csv  = pd.read_csv(ZON_CSV_PATH).dropna(subset=['ZONECLASS'])
zone_feat= pd.read_csv(FEAT_PATH)
zone_emb = pd.read_csv(EMB_PATH)
acs      = pd.read_csv(ACS_PATH)

trips['active_travel'] = trips['mode_type'].isin([1, 2]).astype(int)
# ShapeSTAre_corrected handles split-parcel geometry from spatial joins;
# falls back to original ShapeSTAre if corrected column unavailable.
area_col = ('ShapeSTAre_corrected' if 'ShapeSTAre_corrected' in zon_csv.columns
            else 'ShapeSTAre')
print(f'Area column: {area_col}')

# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# 2. BUILD BLOCK-GROUP FEATURES  (same as classifier scripts)
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
print('\nBuilding block-group features...')

dominant = (zon_csv.sort_values(area_col, ascending=False)
            .drop_duplicates(subset='GEOID', keep='first')[['GEOID','ZONECLASS']]
            .rename(columns={'ZONECLASS':'dominant_zone'}))
dominant['zone_context'] = 'Other/PD'
dominant.loc[dominant['dominant_zone'].isin(RESIDENTIAL),    'zone_context'] = 'Residential'
dominant.loc[dominant['dominant_zone'].isin(NON_RESIDENTIAL),'zone_context'] = 'Non-Residential'

# Area-weighted averages: BG features are area-weighted across zone classes.
# Area-weighted is least lossy vs dominant-zone (ignores minority zones) or
# equal-weight (treats a 1-acre parcel the same as a 100-acre one).
def weighted_mean(group, val_col):
    mask = group[val_col].notna()
    if mask.sum() == 0: return np.nan
    return np.average(group.loc[mask, val_col], weights=group.loc[mask, area_col])

zon_m = zon_csv.merge(
    zone_feat[['zone_class','ped_score_norm','front_setback','lot_width','max_height']],
    left_on='ZONECLASS', right_on='zone_class', how='left')

def entropy_fn(group):
    total = group[area_col].sum()
    if total == 0: return 0.0
    p = group[area_col] / total
    raw = float(-(p[p>0] * np.log(p[p>0])).sum())
    return max(0.0, raw)   # clip numerical floor artifact (e.g. -0.0000)

scraper_agg = (zon_m.groupby('GEOID')
               .apply(lambda g: pd.Series({
                   'ped_score_weighted': weighted_mean(g, 'ped_score_norm'),
                   'front_setback_avg':  weighted_mean(g, 'front_setback'),
                   'lot_width_avg':      weighted_mean(g, 'lot_width'),
                   'max_height_avg':     weighted_mean(g, 'max_height'),
                   'zoning_entropy':     entropy_fn(g),
               }), include_groups=False)
               .reset_index())

# Area-weighted embeddings
emb_cols = [c for c in zone_emb.columns if c.startswith('emb_')]
emb_lookup = zone_emb.set_index('zone_class')[emb_cols].to_dict('index')
bg_emb_rows = []
for geoid, grp in zon_csv.groupby('GEOID'):
    total = grp[area_col].sum()
    vec, cov = np.zeros(len(emb_cols)), 0.0
    for _, row in grp.iterrows():
        z = row['ZONECLASS']
        if z in emb_lookup:
            w = row[area_col] / total
            vec += np.array(list(emb_lookup[z].values())) * w
            cov += w
    if cov > 0:
        bg_emb_rows.append({'GEOID': geoid, **dict(zip(emb_cols, vec))})
bg_emb = pd.DataFrame(bg_emb_rows)

bg = (dominant
      .merge(scraper_agg, on='GEOID', how='left')
      .merge(bg_emb,      on='GEOID', how='left')
      .merge(acs[['GEOID'] + ACS_FEATURES], on='GEOID', how='left'))

df = trips.merge(bg, left_on='o_bg', right_on='GEOID', how='left')
df = df.dropna(subset=['zoning_entropy']).reset_index(drop=True)
print(f'Sample: {len(df):,} trips  |  AT: {df["active_travel"].mean():.1%}')

# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# 3. RESIDUALIZATION FUNCTIONS
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг

def cross_val_r2(X, y, model, n_folds=N_FOLDS):
    """Cross-validated R├ѓ┬▓ via K-fold ├бРѓгРђЮ avoids overfitting bias."""
    kf = KFold(n_splits=n_folds, shuffle=True, random_state=RANDOM_STATE)
    imp = SimpleImputer(strategy='median')
    sc  = StandardScaler()

    y_true_all, y_pred_all = [], []
    for tr_idx, te_idx in kf.split(X):
        X_tr = sc.fit_transform(imp.fit_transform(X[tr_idx]))
        X_te = sc.transform(imp.transform(X[te_idx]))
        y_tr = y[tr_idx]

        m = clone(model)
        m.fit(X_tr, y_tr)
        y_pred_all.extend(m.predict(X_te))
        y_true_all.extend(y[te_idx])

    return r2_score(y_true_all, y_pred_all)


# RF residualization (not OLS) because DML's nuisance model is nonlinear.
# OLS residualization understates demographic entanglement when confounding
# has nonlinear structure -- the opposite of what this diagnostic should show.
def compute_rf_residuals(X, y, n_folds=N_FOLDS):
    """
    Cross-fitted RF residuals: y_res = y - RF_hat(y|X).
    This is the treatment-side nuisance from DML, applied to zoning features.
    Using cross-fitting ensures residuals are out-of-sample predictions,
    matching exactly what DML's nuisance stage computes.
    """
    kf  = KFold(n_splits=n_folds, shuffle=True, random_state=RANDOM_STATE)
    imp = SimpleImputer(strategy='median')
    sc  = StandardScaler()
    rf  = RandomForestRegressor(n_estimators=100, max_depth=6,
                                random_state=RANDOM_STATE, n_jobs=-1)
    y_res = np.zeros(len(y))
    for tr_idx, te_idx in kf.split(X):
        X_tr = sc.fit_transform(imp.fit_transform(X[tr_idx]))
        X_te = sc.transform(imp.transform(X[te_idx]))
        m = clone(rf)
        m.fit(X_tr, y[tr_idx])
        y_res[te_idx] = y[te_idx] - m.predict(X_te)
    return y_res


def compute_ols_residuals(X, y):
    """OLS residuals ├бРѓгРђЮ linear baseline for comparison."""
    imp = SimpleImputer(strategy='median')
    sc  = StandardScaler()
    X_s = sc.fit_transform(imp.fit_transform(X))
    ols = LinearRegression()
    ols.fit(X_s, y)
    return y - ols.predict(X_s)


# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# 4. DIAGNOSTIC TABLE
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
print('\n' + '=' * 65)
print('  PART 1: Zoning-ACS Demographic Predictability Diagnostic')
print('=' * 65)
print('  Framing: R├ѓ┬▓ measures treatment-confounder entanglement.')
print('  It does NOT summarise DML ├бРѓгРђЮ DML depends on both treatment-')
print('  and outcome-side nuisance. Report R├ѓ┬▓ as "demographic')
print('  predictability of the zoning feature" only.\n')

# ACS feature matrix at trip level (one row per trip)
acs_avail = [c for c in ACS_FEATURES if c in df.columns]
X_acs_raw = df[acs_avail].values.astype(float)

# Also load DML ATEs if available
dml_ate = {}
if os.path.exists(DML_PATH):
    dml_df = pd.read_csv(DML_PATH)
    # dml_results.csv has per-embedding-dim results from Estimand B
    # Scalar ATEs (T1/T2/T3) are not in that file ├бРѓгРђЮ hardcode from output
    print('  DML results loaded (embedding dimensions).')
else:
    print(f'  DML results not found at {DML_PATH} ├бРѓгРђЮ ATE column will be empty.')

# Scalar DML ATEs (from theory_dml_acs.py output)
SCALAR_DML_ATES = {
    'ped_score_weighted': {'ate': -0.0195, 'se': 0.0107, 'p': 0.069,  'sig': 'n.s.'},
    'front_setback_avg':  {'ate': -0.0034, 'se': 0.0007, 'p': 0.000,  'sig': '***'},
    # entropy and height were not run as scalar DML treatments
}

diag_rows = []
ols_model = LinearRegression()
rf_model  = RandomForestRegressor(n_estimators=100, max_depth=6,
                                   random_state=RANDOM_STATE, n_jobs=-1)

print(f"  {'Feature':<22} {'OLS R├ѓ┬▓':>7} {'RF R├ѓ┬▓':>7} {'Gap':>7} {'ATE':>9} {'Overlap'}")
print('  ' + '├бРђЮРѓг' * 65)

residuals_store = {}  # store RF residuals for classifier comparison

for feat in ZONING_FEATURES:
    if feat not in df.columns:
        print(f'  {feat:<22}  SKIPPED (not in df)')
        continue

    y_feat = df[feat].values.astype(float)

    # Drop rows where either treatment or ACS is NaN
    valid = ~(np.isnan(y_feat) | np.any(np.isnan(X_acs_raw), axis=1))
    y_v   = y_feat[valid]
    X_v   = X_acs_raw[valid]

    # OLS R├ѓ┬▓
    ols_r2 = cross_val_r2(X_v, y_v, ols_model)
    ols_r2 = max(0.0, ols_r2)   # clip negatives (can occur with CV on small samples)

    # RF R├ѓ┬▓ (cross-validated, matches DML nuisance functional form)
    rf_r2  = cross_val_r2(X_v, y_v, rf_model)
    rf_r2  = max(0.0, rf_r2)

    gap = rf_r2 - ols_r2

    # RF residuals for classifier comparison
    # Compute on full df (imputing NaNs) so trip-level array matches df
    y_all = df[feat].values.astype(float)
    imp   = SimpleImputer(strategy='median')
    y_all_imp = imp.fit_transform(y_all.reshape(-1,1)).ravel()
    res_rf  = compute_rf_residuals(X_acs_raw, y_all_imp)
    res_ols = compute_ols_residuals(X_acs_raw, y_all_imp)
    residuals_store[feat] = {'rf': res_rf, 'ols': res_ols}

    # DML ATE
    dml_info = SCALAR_DML_ATES.get(feat, {})
    ate_str  = (f"{dml_info['ate']:+.4f}{dml_info['sig']}"
                if dml_info else '├бРѓгРђЮ')

    # Overlap flag
    overlap_flag = '├б┼А┬а HIGH' if rf_r2 >= OVERLAP_THRESHOLD else 'OK'

    print(f'  {feat:<22} {ols_r2:>7.3f} {rf_r2:>7.3f} {gap:>+7.3f} {ate_str:>9}  {overlap_flag}')

    diag_rows.append({
        'feature':      feat,
        'ols_r2':       round(ols_r2, 4),
        'rf_r2':        round(rf_r2,  4),
        'ols_rf_gap':   round(gap,    4),
        'dml_ate':      dml_info.get('ate', np.nan),
        'dml_se':       dml_info.get('se',  np.nan),
        'dml_p':        dml_info.get('p',   np.nan),
        'dml_sig':      dml_info.get('sig', ''),
        'overlap_flag': overlap_flag,
    })

diag_df = pd.DataFrame(diag_rows)
diag_df.to_csv('zoning_acs_diagnostic_table.csv', index=False)
print('\n  Saved: zoning_acs_diagnostic_table.csv')

# ├бРђЮРѓг├бРђЮРѓг Interpretation reminder ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
print("""
  ├бРђЮРѓг├бРђЮРѓг Interpretation note ├бРђЮРѓг├бРђЮРѓг
  A high RF R├ѓ┬▓ means the zoning feature is demographically entangled.
  But the null full-sample / significant residential-subgroup pattern
  for ped_score has TWO plausible explanations ├бРѓгРђЮ both should appear
  in the discussion, not just the confounding story:

  (1) CONFOUNDING: ped_score proxies wealthy, educated neighbourhoods.
      After partialling out demographics, the regulatory signal shrinks.
      Supported by: high ACS R├ѓ┬▓ for ped_score.

  (2) EFFECT HETEROGENEITY: ped infrastructure matters more in residential
      zones where walking is viable; in commercial/industrial origins,
      trips are auto-dominated regardless of street design.
      Supported by: the significant residential subgroup ATE in DML,
      and the non-residential ATE being in the opposite direction.

  High R├ѓ┬▓ is consistent with (1) but does NOT rule out (2).
  Low R├ѓ┬▓ would partially rule out (1) but not (2).
  The residualized classifier comparison below adds evidence, but
  the DML subgroup results remain the primary causal identification.
""")

# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# 5. FIGURE 1 ├бРѓгРђЮ OLS R├ѓ┬▓ vs RF R├ѓ┬▓ comparison
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
if len(diag_df) > 0:
    features = diag_df['feature'].str.replace('_', ' ').str.replace('avg', '(avg)').tolist()
    x = np.arange(len(features))
    w = 0.35

    fig1, ax1 = plt.subplots(figsize=(10, 5))
    bars_ols = ax1.bar(x - w/2, diag_df['ols_r2'], w,
                       label='OLS R├ѓ┬▓  (linear confounding)', color='#4878CF', alpha=0.85)
    bars_rf  = ax1.bar(x + w/2, diag_df['rf_r2'],  w,
                       label='RF R├ѓ┬▓   (nonlinear confounding ├бРѓгРђЮ matches DML)',
                       color='#D65F5F', alpha=0.85)

    # Annotate gap where it's large
    for i, row in diag_df.iterrows():
        gap = row['ols_rf_gap']
        if abs(gap) > 0.03:
            ax1.annotate(f'├јРђЮ{gap:+.2f}',
                         xy=(i + w/2, row['rf_r2'] + 0.01),
                         ha='center', fontsize=8, color='#C84B31')

    ax1.axhline(OVERLAP_THRESHOLD, color='orange', linestyle='--', linewidth=1.2,
                label=f'Overlap threshold (RF R├ѓ┬▓ = {OVERLAP_THRESHOLD})')
    ax1.set_xticks(x)
    ax1.set_xticklabels(features, rotation=20, ha='right', fontsize=9)
    ax1.set_ylabel('Cross-validated R├ѓ┬▓\n(ACS ├бРђаРђЎ zoning feature)', fontsize=10)
    ax1.set_title('Demographic Predictability of Zoning Features\n'
                  'OLS R├ѓ┬▓ vs RF R├ѓ┬▓: the gap shows nonlinearity in the confounding structure',
                  fontweight='bold')
    ax1.set_ylim(0, 1.0)
    ax1.legend(fontsize=8)
    ax1.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.savefig('fig_zoning_acs_r2.png', dpi=150, bbox_inches='tight')
    plt.close()
    print('Saved: fig_zoning_acs_r2.png')

# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# 6. OVERLAP DIAGNOSTICS for high-R├ѓ┬▓ features
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
print('\n' + '=' * 65)
print('  PART 2: Overlap / Positivity Diagnostics')
print('=' * 65)

high_r2_features = diag_df[diag_df['rf_r2'] >= OVERLAP_THRESHOLD]['feature'].tolist()
if not high_r2_features:
    print(f'  No features exceed overlap threshold (RF R├ѓ┬▓ ├бРђ░┬Ц {OVERLAP_THRESHOLD}).')
    print('  DML estimates should be stable for all zoning features.')
else:
    print(f'  Features with RF R├ѓ┬▓ ├бРђ░┬Ц {OVERLAP_THRESHOLD}: {high_r2_features}')
    print('  Checking treatment distribution across ACS strata...\n')

    for feat in high_r2_features:
        y_feat = df[feat].values.astype(float)
        # Tertile-split income as a proxy strata
        if 'median_hh_income' in df.columns:
            income = df['median_hh_income'].values
            terciles = np.nanpercentile(income, [33, 67])
            strata = np.where(income < terciles[0], 'Low income',
                     np.where(income < terciles[1], 'Mid income', 'High income'))

            fig_ov, axes = plt.subplots(1, 3, figsize=(12, 4), sharey=False)
            fig_ov.suptitle(
                f'Overlap Diagnostic: {feat}\n'
                f'RF R├ѓ┬▓ = {diag_df.loc[diag_df["feature"]==feat,"rf_r2"].values[0]:.3f} '
                f'├бРѓгРђЮ Treatment distribution across income strata',
                fontweight='bold')
            colors = {'Low income': '#D65F5F', 'Mid income': '#E89C3A', 'High income': '#4878CF'}
            for ax, stratum in zip(axes, ['Low income', 'Mid income', 'High income']):
                vals = y_feat[strata == stratum]
                vals = vals[~np.isnan(vals)]
                ax.hist(vals, bins=25, color=colors[stratum], alpha=0.85, edgecolor='white')
                ax.set_title(f'{stratum}\n(n={len(vals):,})', fontweight='bold')
                ax.set_xlabel(feat.replace('_', ' '))
                ax.set_ylabel('Count')
                ax.grid(alpha=0.3)
            plt.tight_layout()
            fname = f'fig_overlap_{feat}.png'
            plt.savefig(fname, dpi=150, bbox_inches='tight')
            plt.close()
            print(f'  Saved: {fname}')

            # Print overlap summary stats per stratum
            print(f'\n  {feat} distribution by income tertile:')
            print(f"  {'Stratum':<14} {'n':>6} {'mean':>8} {'sd':>8} {'min':>8} {'max':>8}")
            print('  ' + '├бРђЮРѓг' * 56)
            stratum_means = {}
            for stratum in ['Low income', 'Mid income', 'High income']:
                vals = y_feat[strata == stratum]
                vals = vals[~np.isnan(vals)]
                stratum_means[stratum] = np.mean(vals)
                print(f'  {stratum:<14} {len(vals):>6} '
                      f'{np.mean(vals):>8.3f} {np.std(vals):>8.3f} '
                      f'{np.min(vals):>8.3f} {np.max(vals):>8.3f}')

            # U-shape detection: mid-income mean outside [min(low,high), max(low,high)]
            lo, mid, hi = (stratum_means['Low income'],
                           stratum_means['Mid income'],
                           stratum_means['High income'])
            is_u_shape    = mid < min(lo, hi)
            is_inv_u      = mid > max(lo, hi)
            if is_u_shape:
                print(f'\n  ├б┼А┬а U-SHAPE DETECTED for {feat}: mid-income mean ({mid:.3f}) is '
                      f'BELOW both low ({lo:.3f}) and high ({hi:.3f}) strata.')
                print(f'  Likely mechanism: gentrifying urban cores (mid-income, high-entropy/low-setback)')
                print(f'  sandwiched between legacy mixed-use poor and SFR-zoned wealthy suburbs.')
                print(f'  Linear sensitivity analyses (Oster bounds etc.) will miss this entirely.')
                print(f'  The ATE for this feature is effectively identified from within-strata variation;')
                print(f'  mid-income strata carry most of the identification weight.')
            elif is_inv_u:
                print(f'\n  ├б┼А┬а INVERTED-U DETECTED for {feat}: mid-income mean ({mid:.3f}) is '
                      f'ABOVE both low ({lo:.3f}) and high ({hi:.3f}) strata.')

            # Effective sample size: within-strata SD as a share of full-sample SD
            full_sd = np.nanstd(y_feat)
            print(f'\n  Within-strata effective variation (SD / full-sample SD):')
            for stratum in ['Low income', 'Mid income', 'High income']:
                vals = y_feat[strata == stratum]
                vals = vals[~np.isnan(vals)]
                within_sd = np.std(vals)
                pct = within_sd / full_sd if full_sd > 0 else 0
                bar = '├бРђЊ╦є' * int(pct * 20)
                print(f'    {stratum:<14}  {within_sd:.3f}  ({pct:.1%} of full SD)  {bar}')
            print()

        # Also check treatment variance within high-propensity subgroups
        # High RF R├ѓ┬▓ ├бРђаРђЎ some strata have near-deterministic treatment
        # Check: what fraction of observations have RF-predicted value
        # within 10% of the observed range?
        rf_pred_r2 = diag_df.loc[diag_df['feature']==feat, 'rf_r2'].values[0]
        print(f'  RF R├ѓ┬▓ = {rf_pred_r2:.3f}: '
              f'{"├б┼А┬а DML estimates may be unstable in low-variance strata." if rf_pred_r2 > 0.8 else "Moderate overlap concern ├бРѓгРђЮ monitor CIs."}')

# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# 7. RESIDUALIZED CLASSIFIER COMPARISON
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
print('\n' + '=' * 65)
print('  PART 3: Residualized Classifier Comparison')
print('  Residualization method: cross-fitted Random Forest (matches DML)')
print('=' * 65)
print('  Question: does zoning predict active travel beyond demographic sorting?')
print('  Method: replace raw zoning features with RF-residualized versions.')
print('  Gap in RF accuracy = fraction of zoning signal mediated by demographics.\n')

BASE_FEATURES = ['age','gender','education','income_detailed','num_veh',
                 'hhsize','trip_distance','trip_duration','d_purpose_category_imputed']
SCRAPER_FEATURES_RAW = ['zoning_entropy','ped_score_weighted',
                         'front_setback_avg','lot_width_avg','max_height_avg']

# Build residualized scraper features (RF residuals stored above)
for feat, res_dict in residuals_store.items():
    col_name = f'{feat}_res_rf'
    df[col_name] = res_dict['rf']

SCRAPER_FEATURES_RES = [f'{f}_res_rf' for f in SCRAPER_FEATURES_RAW
                         if f'{f}_res_rf' in df.columns]

feat_sets = {
    'A: Baseline':          BASE_FEATURES,
    'B: Raw scraper':       BASE_FEATURES + SCRAPER_FEATURES_RAW,
    'B_res: Resid scraper': BASE_FEATURES + SCRAPER_FEATURES_RES,
}

# Also add residualized embeddings (partial ├бРѓгРђЮ residualize top 5 dims by importance)
top_emb_dims = ['emb_010', 'emb_000', 'emb_004', 'emb_003', 'emb_005']
for dim in top_emb_dims:
    if dim in df.columns:
        y_dim    = df[dim].values.astype(float)
        imp_tmp  = SimpleImputer(strategy='median')
        y_imp    = imp_tmp.fit_transform(y_dim.reshape(-1,1)).ravel()
        df[f'{dim}_res_rf'] = compute_rf_residuals(X_acs_raw, y_imp)

emb_dims_avail     = [d for d in top_emb_dims if d in df.columns]
emb_dims_res_avail = [f'{d}_res_rf' for d in top_emb_dims if f'{d}_res_rf' in df.columns]

if emb_dims_avail:
    feat_sets['C: Raw emb (top 5)']   = BASE_FEATURES + emb_dims_avail
    feat_sets['C_res: Resid emb (top 5)'] = BASE_FEATURES + emb_dims_res_avail

Y    = df['active_travel'].values
# Block-group labels for group-level CV
GEOID_col = 'GEOID' if 'GEOID' in df.columns else 'o_bg'
groups = df[GEOID_col].values if GEOID_col in df.columns else None

clfs = [
    ('Random Forest',     RandomForestClassifier(random_state=RANDOM_STATE)),
    ('Gradient Boosting', GradientBoostingClassifier(random_state=RANDOM_STATE)),
    ('Logistic Reg.',     LogisticRegression(random_state=RANDOM_STATE, max_iter=1000)),
]

# ├бРђЮРѓг├бРђЮРѓг Random-split train/test (original) ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
idx_tr, idx_te = train_test_split(
    np.arange(len(Y)), test_size=0.20, random_state=RANDOM_STATE, stratify=Y)
Y_tr, Y_te = Y[idx_tr], Y[idx_te]

# ├бРђЮРѓг├бРђЮРѓг Repeated 5-fold group-CV by block-group (honest CV) ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# Design: 10 seeds ├ЃРђћ 5 folds = 50 fits total.
#
# Why this structure rather than alternatives:
#   - Single split (n=1): SD uncomputable; result may rest on one lucky fold.
#   - 5 folds (1 seed): 5 fits, SD computable but still seed-dependent.
#   - 50-fold CV: each test fold would contain ~7 BGs ├бРѓгРђЮ too few for stable
#     PR-AUC estimates at 26% base rate.
#   - 10 seeds ├ЃРђћ 5 folds: each of the 50 fits uses ~74 test BGs (20% of 370),
#     keeping individual fold estimates meaningful. SD on the mean reflects
#     genuine fold-assignment variance averaged over seed luck.
#
# The embedding result (C: Raw emb beats baseline on PR-AUC) is the key
# claim in Finding 1. If it holds with PR-AUC SD ├бРђ░┬ц 0.03 across 50 fits,
# the claim is defensible. If SD > 0.03 or mean lift < SD, soften wording.
GROUP_CV_FOLDS = 5
GROUP_CV_SEEDS = 10   # ├бРђаРђЎ 50 fits total
if groups is not None:
    grp_folds = []
    for seed in range(GROUP_CV_SEEDS):
        gss_i = GroupShuffleSplit(n_splits=GROUP_CV_FOLDS,
                                  test_size=0.20,
                                  random_state=RANDOM_STATE + seed)
        for tr_i, te_i in gss_i.split(np.arange(len(Y)), Y, groups):
            grp_folds.append((tr_i, te_i, seed))   # (train_idx, test_idx, seed)
    n_bg_per_fold = [len(np.unique(groups[te])) for _, te, _ in grp_folds]
    print(f'  Repeated group-CV: {GROUP_CV_SEEDS} seeds ├ЃРђћ {GROUP_CV_FOLDS} folds '
          f'= {len(grp_folds)} fits total')
    print(f'  Test BGs per fit: mean={np.mean(n_bg_per_fold):.1f}, '
          f'range {min(n_bg_per_fold)}├бРѓгРђю{max(n_bg_per_fold)}')
else:
    print('  WARNING: GEOID column not found ├бРѓгРђЮ group-CV not available.')
    grp_folds = []

res_clf_rows = []
# Primary metrics: PR-AUC (handles imbalance), AUC (threshold-independent),
# log loss (calibration). Each doing distinct work.
# Accuracy and Brier retained in CSV but treated as supplementary.
print(f"\n  {'Model':<28} {'Classifier':<22} {'PR-AUC':>8} {'AUC':>7} {'Log Loss':>10} {'Acc':>8} {'Brier':>8}")
print('  ' + '├бРђЮРѓг' * 95)
print('  (Primary: PR-AUC / AUC / Log Loss.  Accuracy and Brier supplementary ├бРѓгРђЮ see CSV.)')

for mname, feats in feat_sets.items():
    avail = [f for f in feats if f in df.columns]
    X_raw = df[avail].values.astype(float)
    imp   = SimpleImputer(strategy='median')
    sc    = StandardScaler()

    # ├бРђЮРѓг├бРђЮРѓг Random-split metrics ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
    X_tr  = sc.fit_transform(imp.fit_transform(X_raw[idx_tr]))
    X_te  = sc.transform(imp.transform(X_raw[idx_te]))

    for clf_name, clf_proto in clfs:
        clf = clone(clf_proto)
        clf.fit(X_tr, Y_tr)
        y_pred      = clf.predict(X_te)
        y_prob      = clf.predict_proba(X_te)[:, 1]
        acc         = accuracy_score(Y_te, y_pred)
        auc         = roc_auc_score(Y_te, y_prob)
        pr_auc      = average_precision_score(Y_te, y_prob)
        ll          = log_loss(Y_te, clf.predict_proba(X_te))
        brier       = brier_score_loss(Y_te, y_prob)

        print(f'  {mname:<28} {clf_name:<22} '
              f'{pr_auc:>8.4f} {auc:>7.4f} {ll:>10.4f} {acc:>8.2%} {brier:>8.4f}')

        res_clf_rows.append({
            'model':        mname,
            'classifier':   clf_name,
            'split':        'random',
            'accuracy':     round(acc * 100, 2),
            'accuracy_sd':  np.nan,
            'roc_auc':      round(auc, 4),
            'roc_auc_sd':   np.nan,
            'pr_auc':       round(pr_auc, 4),
            'pr_auc_sd':    np.nan,
            'log_loss':     round(ll, 4),
            'log_loss_sd':  np.nan,
            'brier':        round(brier, 4),
            'residualized': '_res' in mname,
        })

        # ├бРђЮРѓг├бРђЮРѓг Repeated group-CV (RF only, 10 seeds ├ЃРђћ 5 folds = 50 fits) ├бРђЮРѓг├бРђЮРѓг
        # Averages across all 50 fits; SD reflects fold-assignment variance
        # averaged over seed luck, not just within-seed fold variance.
        if grp_folds and clf_name == 'Random Forest':
            fold_metrics = {'pr_auc': [], 'roc_auc': [], 'log_loss': [], 'accuracy': []}
            for fold_tr, fold_te, _seed in grp_folds:
                X_gtr = sc.fit_transform(imp.fit_transform(X_raw[fold_tr]))
                X_gte = sc.transform(imp.transform(X_raw[fold_te]))
                Y_gtr, Y_gte = Y[fold_tr], Y[fold_te]
                clf_g = clone(clf_proto)
                clf_g.fit(X_gtr, Y_gtr)
                gp    = clf_g.predict_proba(X_gte)[:, 1]
                fold_metrics['pr_auc'].append(average_precision_score(Y_gte, gp))
                fold_metrics['roc_auc'].append(roc_auc_score(Y_gte, gp))
                fold_metrics['log_loss'].append(log_loss(Y_gte, clf_g.predict_proba(X_gte)))
                fold_metrics['accuracy'].append(accuracy_score(Y_gte, clf_g.predict(X_gte)))

            n_fits = len(fold_metrics['pr_auc'])
            grp_means = {k: np.mean(v) for k, v in fold_metrics.items()}
            grp_sds   = {k: np.std(v, ddof=1) for k, v in fold_metrics.items()}

            # Summary row: mean across all 50 fits
            res_clf_rows.append({
                'model':        mname,
                'classifier':   clf_name + ' [group-CV mean]',
                'split':        f'group_bg_{GROUP_CV_SEEDS}x{GROUP_CV_FOLDS}_mean',
                'n_fits':       n_fits,
                'accuracy':     round(grp_means['accuracy'] * 100, 2),
                'accuracy_sd':  round(grp_sds['accuracy']  * 100, 3),
                'roc_auc':      round(grp_means['roc_auc'], 4),
                'roc_auc_sd':   round(grp_sds['roc_auc'],  4),
                'pr_auc':       round(grp_means['pr_auc'],  4),
                'pr_auc_sd':    round(grp_sds['pr_auc'],   4),
                'log_loss':     round(grp_means['log_loss'], 4),
                'log_loss_sd':  round(grp_sds['log_loss'],  4),
                'brier':        np.nan,
                'residualized': '_res' in mname,
            })
            # Per-fit rows for traceability (seed and fold encoded in split label)
            for fi, (pr, auc_v, ll_v, acc_v) in enumerate(
                    zip(fold_metrics['pr_auc'], fold_metrics['roc_auc'],
                        fold_metrics['log_loss'], fold_metrics['accuracy'])):
                seed_i = fi // GROUP_CV_FOLDS
                fold_i = fi %  GROUP_CV_FOLDS
                res_clf_rows.append({
                    'model':        mname,
                    'classifier':   f'{clf_name} [group-CV]',
                    'split':        f'group_bg_s{seed_i}_f{fold_i}',
                    'n_fits':       1,
                    'accuracy':     round(acc_v * 100, 2),
                    'accuracy_sd':  np.nan,
                    'roc_auc':      round(auc_v, 4),
                    'roc_auc_sd':   np.nan,
                    'pr_auc':       round(pr,    4),
                    'pr_auc_sd':    np.nan,
                    'log_loss':     round(ll_v,  4),
                    'log_loss_sd':  np.nan,
                    'brier':        np.nan,
                    'residualized': '_res' in mname,
                })
    print()

res_clf_df = pd.DataFrame(res_clf_rows)
res_clf_df.to_csv('residual_classifier_results.csv', index=False)
print('  Saved: residual_classifier_results.csv')

# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# 8. FIGURE 2 ├бРѓгРђЮ Raw vs Residualized: PR-AUC and Log Loss
#    PR-AUC is the primary metric given 26% AT base rate.
#    Log loss is the calibration metric that moves after residualization.
#    Accuracy is in the CSV but not the primary figure.
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
rnd_rf = res_clf_df[
    (res_clf_df['classifier'] == 'Random Forest') &
    (res_clf_df['split'] == 'random')
].copy()

fig2, (ax_pr, ax_ll) = plt.subplots(1, 2, figsize=(14, 5))
fig2.suptitle(
    'Zoning Feature Residualization: PR-AUC and Log Loss (Random Forest, random split)\n'
    'PR-AUC foregrounds imbalanced-outcome performance (26% AT base rate). '
    'Log loss is the metric that detects overconfidence in raw features.',
    fontweight='bold')

for ax, metric, ylabel, ylim_fn in [
    (ax_pr, 'pr_auc',   'PR-AUC',    lambda v: (max(0, min(v)-0.02), min(1, max(v)+0.02))),
    (ax_ll, 'log_loss', 'Log Loss',  lambda v: (max(0, min(v)-0.02), max(v)+0.02)),
]:
    models = rnd_rf['model'].tolist()
    vals   = rnd_rf[metric].tolist()
    colors = ['#A0B8E0' if '_res' in m else '#4878CF' for m in models]
    x      = np.arange(len(models))
    bars   = ax.bar(x, vals, color=colors, alpha=0.85, width=0.6)
    ax.set_xticks(x)
    ax.set_xticklabels([m.replace(': ', ':\n') for m in models],
                       fontsize=7, rotation=15, ha='right')
    lo, hi = ylim_fn(vals)
    ax.set_ylim(lo, hi)
    ax.set_ylabel(ylabel)
    ax.set_title(ylabel, fontweight='bold')
    ax.grid(axis='y', alpha=0.3)
    for bar, v in zip(bars, vals):
        ax.text(bar.get_x() + bar.get_width()/2,
                bar.get_height() + (hi - lo) * 0.01,
                f'{v:.4f}', ha='center', va='bottom', fontsize=7)

raw_patch = mpatches.Patch(color='#4878CF', alpha=0.85, label='Raw features')
res_patch = mpatches.Patch(color='#A0B8E0', alpha=0.85, label='RF-residualized (ACS removed)')
ax_pr.legend(handles=[raw_patch, res_patch], fontsize=8)

plt.tight_layout()
plt.savefig('fig_residual_classifier_comparison.png', dpi=150, bbox_inches='tight')
plt.close()
print('Saved: fig_residual_classifier_comparison.png')

# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# 9. RELIABILITY DIAGRAM (Calibration Plot)
#    Log loss improves after residualization (0.2829├бРђаРђЎ0.2433 for
#    scraper, RF). A reliability diagram checks whether this is
#    consistent with reduced overconfidence (predictions closer
#    to the diagonal) or driven by something else.
#
#    FRAMING NOTE: the diagram can show the improvement is
#    *consistent with* overconfidence driven by demographic-proxy
#    signal ├бРѓгРђЮ it does not *prove* it. Brier barely moves, so the
#    LL improvement reflects a small number of confidently-wrong
#    predictions rather than broad miscalibration. Frame as
#    suggestive, not conclusive.
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
print('\n' + '=' * 65)
print('  PART 4: Reliability Diagrams (Calibration)')
print('  Log loss improves after residualization. Does the')
print('  reliability curve get closer to the diagonal?')
print('  Closer = consistent with overconfidence on demographic proxy.')
print('  Similar = LL improvement driven by other variance reduction.')
print('  (Brier barely moves ├бРђаРђЎ small number of confidently-wrong preds;')
print('   frame as suggestive, not conclusive.)')
print('=' * 65)

rf_proto = RandomForestClassifier(random_state=RANDOM_STATE)
n_bins   = 10
calib_results = {}

for mname, feats in feat_sets.items():
    avail = [f for f in feats if f in df.columns]
    X_raw = df[avail].values.astype(float)
    imp   = SimpleImputer(strategy='median')
    sc    = StandardScaler()
    X_tr  = sc.fit_transform(imp.fit_transform(X_raw[idx_tr]))
    X_te  = sc.transform(imp.transform(X_raw[idx_te]))
    clf   = clone(rf_proto)
    clf.fit(X_tr, Y_tr)
    prob_pos = clf.predict_proba(X_te)[:, 1]
    fraction_of_pos, mean_pred = calibration_curve(Y_te, prob_pos, n_bins=n_bins)
    brier = brier_score_loss(Y_te, prob_pos)
    ll    = log_loss(Y_te, clf.predict_proba(X_te))
    calib_results[mname] = {
        'fraction':  fraction_of_pos,
        'mean_pred': mean_pred,
        'brier':     brier,
        'log_loss':  ll,
    }
    print(f'  {mname:<32}  LL={ll:.4f}  Brier={brier:.4f}')

# ├бРђЮРѓг├бРђЮРѓг Pairing: match 'B: Raw scraper' with 'B_res: Resid scraper' etc.
# Key structure: letter prefix before ':' or '_res:'
# e.g. 'B' ├бРђаРђЎ 'B: Raw scraper' pairs with 'B_res: Resid scraper'
raw_keys = [k for k in calib_results if '_res' not in k]
res_keys = [k for k in calib_results if '_res' in k]
pairs = []
for rk in raw_keys:
    letter = rk.split(':')[0].strip()          # 'A', 'B', 'C'
    matched = [r for r in res_keys
               if r.split(':')[0].strip() == f'{letter}_res']
    if matched:
        pairs.append((rk, matched[0]))

if pairs:
    fig3, axes3 = plt.subplots(1, len(pairs), figsize=(6 * len(pairs), 5))
    if len(pairs) == 1:
        axes3 = [axes3]
    fig3.suptitle(
        'Reliability Diagrams: Raw vs RF-Residualized Features (Random Forest)\n'
        'Closer to diagonal = better calibrated. If residualized curve is closer,\n'
        'this is consistent with (but does not prove) overconfidence driven by '
        'demographic-proxy signal in the raw features.',
        fontweight='bold', fontsize=9)

    for ax, (raw_k, res_k) in zip(axes3, pairs):
        cr   = calib_results[raw_k]
        cres = calib_results[res_k]
        ax.plot([0, 1], [0, 1], 'k--', lw=1, label='Perfect calibration')
        ax.plot(cr['mean_pred'],  cr['fraction'],
                'o-', color='#4878CF', lw=2, ms=5,
                label=f'Raw  (LL={cr["log_loss"]:.4f}, Brier={cr["brier"]:.4f})')
        ax.plot(cres['mean_pred'], cres['fraction'],
                's--', color='#D65F5F', lw=2, ms=5,
                label=f'Residualized  (LL={cres["log_loss"]:.4f}, Brier={cres["brier"]:.4f})')
        ax.set_xlabel('Mean predicted probability', fontsize=10)
        ax.set_ylabel('Fraction of positives (observed AT rate)', fontsize=10)
        ax.set_title(raw_k.split(':')[1].strip(), fontweight='bold')
        ax.legend(fontsize=8)
        ax.set_xlim(0, 1); ax.set_ylim(0, 1)
        ax.grid(alpha=0.3)

        raw_max_dev = np.max(np.abs(cr['fraction']  - cr['mean_pred']))
        res_max_dev = np.max(np.abs(cres['fraction'] - cres['mean_pred']))
        ll_improved = cres['log_loss'] < cr['log_loss']
        brier_moved = abs(cres['brier'] - cr['brier']) > 0.003

        if res_max_dev < raw_max_dev and ll_improved:
            verdict = 'Consistent with demographic overconfidence'
            col = '#2E7D32'
        elif ll_improved and not brier_moved:
            verdict = 'LL improves; Brier flat ├бРђаРђЎ few confident errors'
            col = '#E65100'
        else:
            verdict = 'No clear calibration improvement'
            col = '#C62828'

        ax.text(0.04, 0.91, verdict,
                transform=ax.transAxes, fontsize=7.5,
                color=col, fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.2', fc='white', alpha=0.8))

    plt.tight_layout()
    plt.savefig('fig_reliability_diagrams.png', dpi=150, bbox_inches='tight')
    plt.close()
    print('\nSaved: fig_reliability_diagrams.png')
else:
    print('  (No raw/residualized pairs found ├бРѓгРђЮ check feat_sets key naming)')

# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# 10. GROUP-CV SUMMARY TABLE
#     Compares random-split vs group-split RF accuracy.
#     The gap estimates how much BG-memorisation inflates the
#     random-split numbers. If large ├бРђаРђЎ group-CV is the honest
#     number to report. If small ├бРђаРђЎ random-split numbers hold.
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
print('\n' + '=' * 65)
print('  PART 5: Random-Split vs Group-CV Comparison (RF only)')
print('  Gap = BG-memorisation inflation in random-split accuracy.')
print('  Group-CV folds keep all trips from a given block group')
print('  in the same split, preventing the model from learning')
print('  BG ├бРђаРђЎ AT-rate mappings across the train/test boundary.')
print('=' * 65)

rf_rows  = res_clf_df[res_clf_df['classifier'] == 'Random Forest']
grp_rows = res_clf_df[res_clf_df['classifier'] == 'Random Forest [group-CV]']

grp_rows = res_clf_df[res_clf_df['split'].str.endswith('_mean') &
                      res_clf_df['split'].str.startswith('group_bg')]

if len(grp_rows) > 0:
    print(f"\n  {'Model':<30} {'Rnd PR-AUC':>12} {'Grp PR-AUC (mean├ѓ┬▒SD)':>22} "
          f"{'Grp AUC (mean├ѓ┬▒SD)':>20} {'Gap Acc':>9}")
    print('  ' + '├бРђЮРѓг' * 98)
    for mname in rf_rows['model'].unique():
        r_row = rf_rows[rf_rows['model'] == mname]
        g_row = grp_rows[grp_rows['model'] == mname]
        if len(r_row) == 0 or len(g_row) == 0:
            continue
        r_pr    = r_row['pr_auc'].values[0]
        g_pr    = g_row['pr_auc'].values[0]
        g_pr_sd = g_row['pr_auc_sd'].values[0]
        r_auc   = r_row['roc_auc'].values[0]
        g_auc   = g_row['roc_auc'].values[0]
        g_auc_sd= g_row['roc_auc_sd'].values[0]
        r_acc   = r_row['accuracy'].values[0]
        g_acc   = g_row['accuracy'].values[0]
        acc_gap = r_acc - g_acc

        # Stability check: SD > 0.03 on PR-AUC ├бРђаРђЎ result depends on fold
        pr_stable = g_pr_sd <= 0.03
        stability = '' if pr_stable else '  ├б┼А┬а SD>0.03 UNSTABLE'
        acc_flag  = '  ├б┼А┬а INFLATION' if acc_gap > 1.5 else ''

        print(f'  {mname:<30} {r_pr:>12.4f} '
              f'{g_pr:>10.4f} ├ѓ┬▒ {g_pr_sd:.4f}   '
              f'{g_auc:>8.4f} ├ѓ┬▒ {g_auc_sd:.4f}  '
              f'{acc_gap:>+8.2f}pp{acc_flag}{stability}')

    print()
    print('  Stability interpretation (50-fit repeated group-CV):')
    print('  PR-AUC SD ├бРђ░┬ц 0.03 ├бРђаРђЎ result is stable across fold assignments.')
    print('  PR-AUC SD > 0.03 ├бРђаРђЎ meaningful fold-assignment variance; soften wording.')
    print('  SD here reflects variance across all 50 fits (10 seeds ├ЃРђћ 5 folds),')
    print('  not just within-seed variance ├бРѓгРђЮ harder to dismiss as a single lucky fold.')
    print()
    print('  The embedding result (C: Raw emb) is the key claim in Finding 1.')
    print('  If C beats A (Baseline) with mean lift > 1 SD, the claim is defensible.')
    print('  Individual fit results in CSV (split = group_bg_sN_fM rows).')
    print()

    # Print fold-level detail for the embedding models specifically
    emb_model_names = [m for m in res_clf_df['model'].unique() if 'emb' in m.lower()]
    fold_detail = res_clf_df[
        res_clf_df['split'].str.startswith('group_bg_s') &
        res_clf_df['model'].isin(['A: Baseline'] + emb_model_names)
    ][['model', 'split', 'pr_auc', 'roc_auc', 'log_loss']].sort_values(['model', 'split'])

    if len(fold_detail) > 0:
        print('  Per-fold PR-AUC detail (Baseline vs Embeddings ├бРѓгРђЮ the key comparison):')
        print(f"  {'Model':<30} {'Fold':<20} {'PR-AUC':>8} {'AUC':>8}")
        print('  ' + '├бРђЮРѓг' * 70)
        for _, row in fold_detail.iterrows():
            print(f"  {row['model']:<30} {row['split']:<20} "
                  f"{row['pr_auc']:>8.4f} {row['roc_auc']:>8.4f}")
else:
    print('  Group-CV results not available (GEOID column not found in df).')

# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# 12. CONTEXT-SPLIT GROUP-CV
#     Repeats the 50-fit repeated group-CV for residential and
#     non-residential origin subsamples separately.
#
#     Key claims that need checking:
#       Residential:     Model B (+scraper) leads at 92├бРѓгРђю94% RF acc
#                        C1/C2 embedding tie at 94.00%
#                        ACS (E) adds nothing over baseline
#       Non-residential: Model E (+ACS) ties Model B at 89.08%
#                        Embeddings tie scraper in non-res
#
#     If these hold under group-CV, they survive BG-memorisation.
#     If they don't, the context-split narrative needs softening.
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
print('\n' + '=' * 65)
print('  PART 6: Context-Split Repeated Group-CV')
print(f'  {GROUP_CV_SEEDS} seeds ├ЃРђћ {GROUP_CV_FOLDS} folds = '
      f'{GROUP_CV_SEEDS * GROUP_CV_FOLDS} fits per context')
print('=' * 65)

# Context masks
RES_LABEL    = 'Residential'
NONRES_LABEL = 'Non-Residential'
ctx_masks = {}
if 'zone_context' in df.columns:
    ctx_masks[RES_LABEL]    = (df['zone_context'] == RES_LABEL).values
    ctx_masks[NONRES_LABEL] = (df['zone_context'] == NONRES_LABEL).values
else:
    # Fall back to dominant_zone membership if zone_context not present
    res_zones = [
        'RS-150','RS-100','RS-75','RS-60','RS-50',
        'RM-12','RM-16','RM-18','RM-24','RM-35','RM-50','RM-75',
        'RO','RO-1','SH-RS','SH-RS-A','SH-RM','SH-RO','SH-PD',
        'YC-2','YC-4','YC-8','YC-9',
    ]
    if 'dominant_zone' in df.columns:
        ctx_masks[RES_LABEL]    = df['dominant_zone'].isin(res_zones).values
        ctx_masks[NONRES_LABEL] = ~df['dominant_zone'].isin(res_zones).values
    else:
        print('  WARNING: zone_context and dominant_zone not found ├бРѓгРђЮ skipping context split.')
        ctx_masks = {}

# Models to test in context split ├бРѓгРђЮ key comparisons only
ctx_feat_sets = {k: v for k, v in feat_sets.items()
                 if k in ['A: Baseline', 'B: Raw scraper',
                           'C: Raw emb (top 5)', 'E: +ACS']
                 or 'E' in k}
# If ACS model not built yet (v9-style E), add it
ACS_FEATURES_AVAIL = [c for c in ACS_FEATURES if c in df.columns]
if 'E: +ACS' not in feat_sets and ACS_FEATURES_AVAIL:
    ctx_feat_sets['E: +ACS'] = BASE_FEATURES + ACS_FEATURES_AVAIL

ctx_clf_rows = []

for ctx_label, mask in ctx_masks.items():
    n_ctx   = mask.sum()
    at_rate = df.loc[mask, 'active_travel'].mean()
    print(f'\n  Context: {ctx_label}  (n={n_ctx:,}, AT={at_rate:.1%})')

    Y_ctx    = Y[mask]
    grps_ctx = groups[mask] if groups is not None else None

    if grps_ctx is None:
        print('    GEOID not available ├бРѓгРђЮ skipping.')
        continue

    n_bg_ctx = len(np.unique(grps_ctx))
    print(f'  Block groups: {n_bg_ctx}')

    # Build group-CV folds for this context
    ctx_folds = []
    for seed in range(GROUP_CV_SEEDS):
        gss_c = GroupShuffleSplit(n_splits=GROUP_CV_FOLDS,
                                  test_size=0.20,
                                  random_state=RANDOM_STATE + seed)
        for tr_i, te_i in gss_c.split(np.arange(n_ctx), Y_ctx, grps_ctx):
            ctx_folds.append((tr_i, te_i, seed))

    n_bg_test_ctx = [len(np.unique(grps_ctx[te])) for _, te, _ in ctx_folds]
    print(f'  Folds: {len(ctx_folds)} fits, '
          f'~{np.mean(n_bg_test_ctx):.1f} test BGs per fit')

    # Random-split baseline for this context (for comparison)
    idx_ctx_tr, idx_ctx_te = train_test_split(
        np.arange(n_ctx), test_size=0.20,
        random_state=RANDOM_STATE, stratify=Y_ctx)

    print(f"\n    {'Model':<28} {'Rnd PR-AUC':>12} {'Grp PR-AUC (mean├ѓ┬▒SD)':>22} "
          f"{'Grp AUC (mean├ѓ┬▒SD)':>20} {'Grp LL (mean├ѓ┬▒SD)':>20}")
    print('    ' + '├бРђЮРѓг' * 100)

    for mname, feats in ctx_feat_sets.items():
        avail   = [f for f in feats if f in df.columns]
        X_ctx   = df.loc[mask, avail].values.astype(float)

        # Random split
        imp_c = SimpleImputer(strategy='median')
        sc_c  = StandardScaler()
        X_ctr = sc_c.fit_transform(imp_c.fit_transform(X_ctx[idx_ctx_tr]))
        X_cte = sc_c.transform(imp_c.transform(X_ctx[idx_ctx_te]))
        clf_r = clone(RandomForestClassifier(random_state=RANDOM_STATE))
        clf_r.fit(X_ctr, Y_ctx[idx_ctx_tr])
        rnd_prob = clf_r.predict_proba(X_cte)[:, 1]
        rnd_pr   = average_precision_score(Y_ctx[idx_ctx_te], rnd_prob)

        # Repeated group-CV
        fold_pr, fold_auc, fold_ll = [], [], []
        for fold_tr, fold_te, _s in ctx_folds:
            X_gtr = sc_c.fit_transform(imp_c.fit_transform(X_ctx[fold_tr]))
            X_gte = sc_c.transform(imp_c.transform(X_ctx[fold_te]))
            Y_gtr, Y_gte = Y_ctx[fold_tr], Y_ctx[fold_te]
            # Skip folds with only one class in test ├бРѓгРђЮ can happen in small contexts
            if len(np.unique(Y_gte)) < 2:
                continue
            clf_g = clone(RandomForestClassifier(random_state=RANDOM_STATE))
            clf_g.fit(X_gtr, Y_gtr)
            gp = clf_g.predict_proba(X_gte)[:, 1]
            fold_pr.append(average_precision_score(Y_gte, gp))
            fold_auc.append(roc_auc_score(Y_gte, gp))
            fold_ll.append(log_loss(Y_gte, clf_g.predict_proba(X_gte)))

        if not fold_pr:
            print(f'    {mname:<28}  (no valid folds)')
            continue

        m_pr  = np.mean(fold_pr);  sd_pr  = np.std(fold_pr,  ddof=1)
        m_auc = np.mean(fold_auc); sd_auc = np.std(fold_auc, ddof=1)
        m_ll  = np.mean(fold_ll);  sd_ll  = np.std(fold_ll,  ddof=1)

        # Flag: does this model beat baseline by more than 1 SD?
        flag = ''
        if mname != 'A: Baseline':
            # Get baseline mean for this context (already computed if available)
            base_rows = [r for r in ctx_clf_rows
                         if r['context'] == ctx_label and r['model'] == 'A: Baseline']
            if base_rows:
                base_pr = base_rows[0]['grp_pr_auc_mean']
                base_sd = base_rows[0]['grp_pr_auc_sd']
                lift = m_pr - base_pr
                if lift > base_sd:
                    flag = '  ├б┼ЊРђю beats baseline > 1 SD'
                elif lift > 0:
                    flag = '  ~ lift within 1 SD'
                else:
                    flag = '  ├б┼ЊРђћ does not beat baseline'

        print(f'    {mname:<28} {rnd_pr:>12.4f} '
              f'{m_pr:>10.4f} ├ѓ┬▒ {sd_pr:.4f}   '
              f'{m_auc:>8.4f} ├ѓ┬▒ {sd_auc:.4f}  '
              f'{m_ll:>8.4f} ├ѓ┬▒ {sd_ll:.4f}{flag}')

        ctx_clf_rows.append({
            'context':           ctx_label,
            'model':             mname,
            'n_trips':           n_ctx,
            'at_rate':           round(at_rate, 4),
            'n_fits':            len(fold_pr),
            'rnd_pr_auc':        round(rnd_pr,  4),
            'grp_pr_auc_mean':   round(m_pr,    4),
            'grp_pr_auc_sd':     round(sd_pr,   4),
            'grp_auc_mean':      round(m_auc,   4),
            'grp_auc_sd':        round(sd_auc,  4),
            'grp_ll_mean':       round(m_ll,    4),
            'grp_ll_sd':         round(sd_ll,   4),
        })

ctx_clf_df = pd.DataFrame(ctx_clf_rows)
if len(ctx_clf_df) > 0:
    ctx_clf_df.to_csv('context_split_groupcv_results.csv', index=False)
    print('\n  Saved: context_split_groupcv_results.csv')

    # Figure: PR-AUC mean ├ѓ┬▒ SD by model and context
    if len(ctx_masks) == 2:
        models_ctx = ctx_clf_df['model'].unique()
        contexts   = [RES_LABEL, NONRES_LABEL]
        x          = np.arange(len(models_ctx))
        width      = 0.35
        colors_ctx = {RES_LABEL: '#4878CF', NONRES_LABEL: '#D65F5F'}

        fig_ctx, ax_ctx = plt.subplots(figsize=(12, 5))
        for ci, ctx_label in enumerate(contexts):
            sub  = ctx_clf_df[ctx_clf_df['context'] == ctx_label]
            # Align to models_ctx order
            means = [sub.loc[sub['model']==m, 'grp_pr_auc_mean'].values[0]
                     if len(sub[sub['model']==m]) > 0 else np.nan
                     for m in models_ctx]
            sds   = [sub.loc[sub['model']==m, 'grp_pr_auc_sd'].values[0]
                     if len(sub[sub['model']==m]) > 0 else np.nan
                     for m in models_ctx]
            offset = (ci - 0.5) * width
            ax_ctx.bar(x + offset, means, width * 0.9,
                       yerr=sds, capsize=4,
                       color=colors_ctx[ctx_label], alpha=0.85,
                       label=ctx_label, error_kw={'elinewidth': 1.5})

        ax_ctx.set_xticks(x)
        ax_ctx.set_xticklabels([m.replace(': ', ':\n') for m in models_ctx],
                                fontsize=9, rotation=10, ha='right')
        ax_ctx.set_ylabel('PR-AUC (mean ├ѓ┬▒ SD, 50-fit group-CV)')
        ax_ctx.set_title(
            'Context-Split PR-AUC: Residential vs Non-Residential Origins\n'
            f'Repeated group-CV ({GROUP_CV_SEEDS} seeds ├ЃРђћ {GROUP_CV_FOLDS} folds). '
            'Error bars = SD across 50 fits.',
            fontweight='bold')
        ax_ctx.legend()
        ax_ctx.grid(axis='y', alpha=0.3)
        plt.tight_layout()
        plt.savefig('fig_context_split_groupcv.png', dpi=150, bbox_inches='tight')
        plt.close()
        print('  Saved: fig_context_split_groupcv.png')

    print('\n  Context-split interpretation notes:')
    print('  Residential ├бРѓгРђЮ key claims to verify:')
    print('    B (+scraper) should beat A (baseline) with lift > 1 SD')
    print('    C (embeddings) should tie or beat B in group-CV PR-AUC')
    print('    E (+ACS) should add little over A in residential')
    print('  Non-residential ├бРѓгРђЮ key claims to verify:')
    print('    E (+ACS) should tie B in group-CV PR-AUC')
    print('    C (embeddings) should tie B in group-CV PR-AUC')
    print('    If neither B nor E beats A under group-CV ├бРђаРђЎ all context-split')
    print('    gains were BG-memorisation; context-split narrative needs revision.')

# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
# 13. FINAL SUMMARY
# ├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг├бРђЮРѓг
print('\n' + '=' * 65)
print('  SUMMARY ├бРѓгРђЮ Diagnostic Chain')
print('=' * 65)


if len(diag_df) > 0:
    print('  Entanglement summary (RF R├ѓ┬▓):')
    for _, row in diag_df.iterrows():
        flag = '├б┼А┬а' if row['rf_r2'] >= OVERLAP_THRESHOLD else ' '
        print(f'  {flag} {row["feature"]:<24}  OLS={row["ols_r2"]:.3f}  '
              f'RF={row["rf_r2"]:.3f}  gap={row["ols_rf_gap"]:+.3f}  '
              f'ATE={row["dml_ate"] if not np.isnan(row["dml_ate"]) else "├бРѓгРђЮ"}')

print('\n  Primary metrics (PR-AUC / AUC / LL), RF, random split:')
rf_results = res_clf_df[(res_clf_df['classifier']=='Random Forest') &
                         (res_clf_df['split']=='random')]
for _, row in rf_results.iterrows():
    marker = '├бРђа┬љresidualized' if row['residualized'] else ''
    print(f'    {row["model"]:<30}  PR-AUC={row["pr_auc"]:.4f}  '
          f'AUC={row["roc_auc"]:.4f}  LL={row["log_loss"]:.4f}  '
          f'(Acc={row["accuracy"]:.2f}%)  {marker}')




print('\n├б┼ЊРђю Residualization diagnostic complete.')
print('\nOutputs:')
print('  zoning_acs_diagnostic_table.csv        ├бРђа┬љ entanglement table (OLS R├ѓ┬▓, RF R├ѓ┬▓, gap, ATE)')
print('  fig_zoning_acs_r2.png                  ├бРђа┬љ OLS vs RF R├ѓ┬▓ bar chart')
print('  fig_overlap_<feature>.png              ├бРђа┬љ treatment distributions + U-shape detection')
print('  residual_classifier_results.csv        ├бРђа┬љ PR-AUC / AUC / LL / Acc / Brier,')
print('                                            random-split and group-CV rows (50 fits)')
print('  fig_residual_classifier_comparison.png ├бРђа┬љ PR-AUC and LL panels (primary figure)')
print('  fig_reliability_diagrams.png           ├бРђа┬љ calibration plots raw vs residualized')
print('  context_split_groupcv_results.csv      ├бРђа┬љ residential/non-res group-CV mean ├ѓ┬▒ SD')
print('  fig_context_split_groupcv.png          ├бРђа┬љ PR-AUC bar chart by context')
```
