```python
"""
ACS Block-Group Sociodemographic Data — Tampa / Hillsborough County
====================================================================
Downloads 2019 ACS 5-year estimates at the census block group level
for Hillsborough County, FL (FIPS 12-057) and produces a clean CSV
ready to merge into the active travel analysis pipeline.

Year rationale:
  Travel survey ran Oct 2018 – May 2019. ACS 5-year 2019 (covering
  2015–2019) is the closest contemporaneous estimate available at the
  block group level. ACS 1-year data does not go below tract level,
  so 5-year is the only option at this geography — standard in the
  built environment / travel behavior literature.

Setup:
  pip install census
  Get a free API key at: api.census.gov/data/key_signup.html
  Set via environment variable: export CENSUS_API_KEY=your_key

Output:
  acs_blockgroup_tampa.csv   (one row per block group, 12-digit GEOID)

Columns produced:
  GEOID                  12-digit integer, matches o_bg in tampatrips_1_.csv
  median_hh_income       Median household income ($); NaN where unavailable
  total_pop              Total population
  total_hh               Total households (vehicle availability denominator)
  hh_no_vehicle          Households with no vehicle available
  pct_no_vehicle         hh_no_vehicle / total_hh
  educ_total             Population 25+ (education denominator)
  educ_bachelors_plus    Bachelor's + Master's + Professional + Doctorate
  pct_bachelors_plus     educ_bachelors_plus / educ_total
  total_race             Total population (race denominator)
  white_nonhisp          White non-Hispanic alone
  black_nonhisp          Black or African American non-Hispanic alone
  hispanic               Hispanic or Latino (any race)
  pct_white_nonhisp      white_nonhisp / total_race
  pct_black_nonhisp      black_nonhisp / total_race
  pct_hispanic           hispanic / total_race
  poverty_total          Families (poverty denominator)
  poverty_below          Families below poverty level
  pct_poverty            poverty_below / poverty_total
  housing_total          Total occupied housing units
  owner_occupied         Owner-occupied housing units
  pct_owner_occ          owner_occupied / housing_total

Usage in DML pipeline:
  Merge on GEOID (or o_bg), add ACS columns to CONFOUNDERS_FULL.
  BG-level controls capture neighbourhood-level residential self-selection
  that individual survey demographics alone cannot fully absorb.
"""

import os
import numpy as np
import pandas as pd
from census import Census

# ─────────────────────────────────────────────────────────────
# CONFIGURATION
# ─────────────────────────────────────────────────────────────
API_KEY = os.environ['CENSUS_API_KEY']  # set via: export CENSUS_API_KEY=your_key

STATE_FIPS  = '12'    # Florida
COUNTY_FIPS = '057'   # Hillsborough County
ACS_YEAR    = 2019    # 5-year ACS 2019 (covers 2015–2019)

OUTPUT_PATH = 'acs_blockgroup_tampa.csv'

# Census sentinel values for missing/suppressed data
# These appear as large negative integers and must be treated as NaN
CENSUS_NULLS = [-666666666, -999999999, -888888888]

# ─────────────────────────────────────────────────────────────
# VARIABLE DEFINITIONS
# ─────────────────────────────────────────────────────────────
# Maps Census variable codes -> friendly column names
VARIABLES = {
    # Income
    'B19013_001E': 'median_hh_income',

    # Population
    'B01003_001E': 'total_pop',

    # Vehicle availability (B08201: households by number of vehicles)
    'B08201_001E': 'total_hh',
    'B08201_002E': 'hh_no_vehicle',

    # Educational attainment for population 25+ (B15003)
    'B15003_001E': 'educ_total',
    'B15003_022E': 'educ_bachelors',
    'B15003_023E': 'educ_masters',
    'B15003_024E': 'educ_professional',
    'B15003_025E': 'educ_doctorate',

    # Race and ethnicity (B03002: Hispanic/Latino origin by race)
    'B03002_001E': 'total_race',
    'B03002_003E': 'white_nonhisp',
    'B03002_004E': 'black_nonhisp',
    'B03002_012E': 'hispanic',

    # Poverty status (B17010: poverty status of families)
    'B17010_001E': 'poverty_total',
    'B17010_002E': 'poverty_below',

    # Housing tenure (B25003: tenure)
    'B25003_001E': 'housing_total',
    'B25003_002E': 'owner_occupied',
}

# ─────────────────────────────────────────────────────────────
# 1. FETCH FROM API
# ─────────────────────────────────────────────────────────────
print("=" * 60)
print("  ACS Block-Group Data — Hillsborough County, FL")
print(f"  ACS 5-year {ACS_YEAR}")
print("=" * 60)

c = Census(API_KEY)

print("\nFetching from Census API...")
raw = c.acs5.state_county_blockgroup(
    fields=['GEO_ID'] + list(VARIABLES.keys()),
    state_fips=STATE_FIPS,
    county_fips=COUNTY_FIPS,
    tract='*',
    blockgroup='*',
    year=ACS_YEAR,
)

df = pd.DataFrame(raw)
print(f"  Rows returned: {len(df)}")
print(f"  Columns: {df.columns.tolist()}")

# ─────────────────────────────────────────────────────────────
# 2. RENAME AND CLEAN
# ─────────────────────────────────────────────────────────────

# Rename to friendly names
df = df.rename(columns=VARIABLES)

# Cast all numeric columns from string/object to numeric
# (Census API returns everything as strings)
numeric_cols = list(VARIABLES.values())
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Replace Census sentinel null values with NaN
# Must happen AFTER numeric conversion, before any rate calculations
df[numeric_cols] = df[numeric_cols].replace(CENSUS_NULLS, np.nan)

print(f"\nNull counts after sentinel replacement:")
for col in numeric_cols:
    n = df[col].isnull().sum()
    if n > 0:
        print(f"  {col:<28} {n} nulls")

# ─────────────────────────────────────────────────────────────
# 3. BUILD GEOID
#    Census API returns geographic identifiers as separate columns:
#    'state' (2 chars), 'county' (3 chars), 'tract' (6 chars, zero-padded),
#    'block group' (1 char).
#    Concatenated and cast to int gives a 12-digit integer matching
#    the GEOID format in tpazoning_clean.csv and tampatrips_1_.csv.
# ─────────────────────────────────────────────────────────────
df['GEOID'] = (
    df['state'].str.zfill(2) +
    df['county'].str.zfill(3) +
    df['tract'].str.zfill(6) +    # Census API returns 6-char already, but zfill is safe
    df['block group'].str.zfill(1)
).astype(int)

# Sanity check: all GEOIDs should be 12 digits
geoid_lengths = df['GEOID'].astype(str).str.len()
if not (geoid_lengths == 12).all():
    bad = df[geoid_lengths != 12][['GEOID','state','county','tract','block group']]
    print(f"\nWARNING: {len(bad)} GEOIDs are not 12 digits:")
    print(bad.head())
else:
    print(f"\nGEOID check passed: all {len(df)} GEOIDs are 12 digits")

# ─────────────────────────────────────────────────────────────
# 4. DERIVED RATES
#    Rates are more meaningful as model inputs than raw counts,
#    and are robust to BG population size variation.
#    Zero denominators -> NaN (SimpleImputer handles these downstream).
# ─────────────────────────────────────────────────────────────

def safe_rate(num, denom):
    """Divide num by denom, returning NaN where denom is zero or NaN."""
    return np.where(
        (denom > 0) & denom.notna() & num.notna(),
        num / denom,
        np.nan
    )

df['educ_bachelors_plus'] = (df['educ_bachelors'] + df['educ_masters'] +
                              df['educ_professional'] + df['educ_doctorate'])

df['pct_no_vehicle']      = safe_rate(df['hh_no_vehicle'],       df['total_hh'])
df['pct_bachelors_plus']  = safe_rate(df['educ_bachelors_plus'], df['educ_total'])
df['pct_white_nonhisp']   = safe_rate(df['white_nonhisp'],       df['total_race'])
df['pct_black_nonhisp']   = safe_rate(df['black_nonhisp'],       df['total_race'])
df['pct_hispanic']        = safe_rate(df['hispanic'],            df['total_race'])
df['pct_poverty']         = safe_rate(df['poverty_below'],       df['poverty_total'])
df['pct_owner_occ']       = safe_rate(df['owner_occupied'],      df['housing_total'])

# Rate sanity checks: all rates should be in [0, 1]
rate_cols = ['pct_no_vehicle', 'pct_bachelors_plus', 'pct_white_nonhisp',
             'pct_black_nonhisp', 'pct_hispanic', 'pct_poverty', 'pct_owner_occ']
print("\nRate range checks (should all be 0-1):")
for col in rate_cols:
    lo  = df[col].min()
    hi  = df[col].max()
    bad = ((df[col] < 0) | (df[col] > 1)).sum()
    flag = " <-- CHECK" if bad > 0 else ""
    print(f"  {col:<26} [{lo:.3f}, {hi:.3f}]  {bad} out-of-range{flag}")

# ─────────────────────────────────────────────────────────────
# 5. SELECT OUTPUT COLUMNS AND SAVE
# ─────────────────────────────────────────────────────────────
OUTPUT_COLS = [
    'GEOID',
    # Raw counts kept for reference / alternative denominators
    'median_hh_income',
    'total_pop',
    'total_hh',
    'hh_no_vehicle',
    'educ_total',
    'educ_bachelors_plus',
    'total_race',
    'white_nonhisp',
    'black_nonhisp',
    'hispanic',
    'poverty_total',
    'poverty_below',
    'housing_total',
    'owner_occupied',
    # Rates — these are the primary inputs for CONFOUNDERS_FULL
    'pct_no_vehicle',
    'pct_bachelors_plus',
    'pct_white_nonhisp',
    'pct_black_nonhisp',
    'pct_hispanic',
    'pct_poverty',
    'pct_owner_occ',
]

out = df[OUTPUT_COLS].copy()
out.to_csv(OUTPUT_PATH, index=False)
print(f"\nSaved: {OUTPUT_PATH}  ({len(out)} block groups, {len(out.columns)} columns)")
```
