```python
"""
Tampa Zoning CSV — Split Parcel Fix
=====================================
Problem: 1,362 zoning parcels (35% of unique parcels, 59.5% of rows) cross
census block group boundaries. The spatial join that produced tpazoning.csv
duplicated these parcels — one row per block group touched — but kept the
full unclipped parcel area (ShapeSTAre) in every row instead of the true
partial area for each block group.

This inflated the apparent area of large zone classes in many block groups.
302 out of 370 block groups (82%) had the wrong dominant zone class as a result.

Fix: The geometry column contains the actual clipped intersection polygon for
each OBJECTID × GEOID combination. We compute the area of each intersection
polygon using the shoelace formula and use that as the corrected area weight.
For non-split parcels the corrected and original areas are proportional (r=0.9997).

Outputs:
  tpazoning_clean.csv        — corrected file, drop-in replacement for tpazoning.csv
  split_parcel_report.csv    — summary of which parcels were split and how
"""

import re
import numpy as np
import pandas as pd

# ─────────────────────────────────────────────────────────────
# 1. LOAD
# ─────────────────────────────────────────────────────────────
df = pd.read_csv('tpazoning.csv').dropna(subset=['ZONECLASS'])
print(f"Loaded: {len(df):,} rows, {df['OBJECTID'].nunique():,} unique parcels")

# ─────────────────────────────────────────────────────────────
# 2. COMPUTE INTERSECTION AREA FROM GEOMETRY
#    Shoelace formula on the first exterior ring of each WKT polygon.
#    Result is in Web Mercator coordinate units — scale doesn't matter
#    since we only use relative proportions for area-weighting.
# ─────────────────────────────────────────────────────────────
def geom_area(geom_str):
    """
    Area of the first exterior ring via the shoelace formula.
    Works for POLYGON and MULTIPOLYGON WKT (uses first ring only).
    Returns 0 for null or degenerate geometries.
    """
    if not isinstance(geom_str, str):
        return 0.0
    coords = re.findall(
        r'(-?\d+\.?\d+(?:e[+-]?\d+)?)\s+(-?\d+\.?\d+(?:e[+-]?\d+)?)',
        geom_str
    )
    if len(coords) < 3:
        return 0.0
    xs = [float(c[0]) for c in coords]
    ys = [float(c[1]) for c in coords]
    n  = len(xs)
    return abs(sum(
        xs[i] * ys[(i+1) % n] - xs[(i+1) % n] * ys[i]
        for i in range(n)
    )) / 2.0

print("Computing intersection areas from geometry column...")
df['area_intersect'] = df['geometry'].apply(geom_area)

# Fallback: for any row where geometry is missing/degenerate, use ShapeSTAre
# proportionally scaled so non-split parcel areas match their ShapeSTAre
zero_mask = df['area_intersect'] == 0
if zero_mask.sum() > 0:
    print(f"  {zero_mask.sum()} rows with zero computed area — "
          f"falling back to ShapeSTAre for those rows")
    df.loc[zero_mask, 'area_intersect'] = df.loc[zero_mask, 'ShapeSTAre']

# ─────────────────────────────────────────────────────────────
# 3. REPORT SPLIT PARCELS
# ─────────────────────────────────────────────────────────────
obj_geoid_count = df.groupby('OBJECTID')['GEOID'].nunique().rename('n_geoids')
df = df.merge(obj_geoid_count, on='OBJECTID')

split_df    = df[df['n_geoids'] > 1]
nonsplit_df = df[df['n_geoids'] == 1]

print(f"\nSplit parcel summary:")
print(f"  Split parcels (cross BG boundary): {split_df['OBJECTID'].nunique():,} "
      f"({split_df['OBJECTID'].nunique()/df['OBJECTID'].nunique():.1%})")
print(f"  Rows from split parcels:           {len(split_df):,} "
      f"({len(split_df)/len(df):.1%})")
print(f"  Non-split parcels:                 {nonsplit_df['OBJECTID'].nunique():,}")

# Save split parcel report
report = (split_df.groupby('OBJECTID')
          .agg(
              ZONECLASS    = ('ZONECLASS', 'first'),
              n_geoids     = ('GEOID', 'nunique'),
              geoids       = ('GEOID', lambda x: '|'.join(x.astype(str).unique())),
              ShapeSTAre   = ('ShapeSTAre', 'first'),
              total_intersect_area = ('area_intersect', 'sum'),
          )
          .reset_index()
          .sort_values('n_geoids', ascending=False))
report.to_csv('split_parcel_report.csv', index=False)
print("Saved: split_parcel_report.csv")

# ─────────────────────────────────────────────────────────────
# 4. VALIDATE AREA CORRECTION
# ─────────────────────────────────────────────────────────────
# For non-split parcels, computed area should be proportional to ShapeSTAre
corr = nonsplit_df['area_intersect'].corr(nonsplit_df['ShapeSTAre'])
print(f"\nValidation — correlation of computed vs. original area "
      f"(non-split parcels): r = {corr:.4f}")

# For split parcels, sum of intersection areas should approximate ShapeSTAre
split_check = (split_df.groupby('OBJECTID')
               .agg(ShapeSTAre=('ShapeSTAre','first'),
                    sum_intersect=('area_intersect','sum'))
               .assign(ratio=lambda x: x['sum_intersect'] / x['ShapeSTAre']))
# Scale factor between coordinate units and sq feet
scale = split_check['ratio'].median()
print(f"Median scale factor (coord units / ShapeSTAre): {scale:.6f}")

# ─────────────────────────────────────────────────────────────
# 5. WRITE CORRECTED FILE
#    Replace ShapeSTAre with area_intersect (scaled to sq feet
#    using the median scale factor from non-split parcels).
#    This gives each OBJECTID × GEOID row its true partial area.
# ─────────────────────────────────────────────────────────────
# Scale intersection areas to approximate sq feet
nonsplit_scale = (nonsplit_df['ShapeSTAre'] / nonsplit_df['area_intersect']
                  .replace(0, np.nan)).median()
df['ShapeSTAre_corrected'] = df['area_intersect'] * nonsplit_scale

# Write clean file with corrected area column
# Keep original ShapeSTAre for reference, add ShapeSTAre_corrected
clean_cols = [c for c in df.columns if c not in ['area_intersect','n_geoids',
                                                        'ShapeSTAre_corrected']]
df_clean = df[clean_cols + ['ShapeSTAre_corrected']].copy()
df_clean.to_csv('tpazoning_clean.csv', index=False)

print(f"\nSaved: tpazoning_clean.csv  ({len(df_clean):,} rows)")
print("  ShapeSTAre          = original full-parcel area (for reference)")
print("  ShapeSTAre_corrected = intersection area scaled to sq feet (USE THIS)")

# ─────────────────────────────────────────────────────────────
# 6. VERIFY DOMINANT ZONE CHANGES
# ─────────────────────────────────────────────────────────────
def get_dominant(frame, area_col):
    # Sort descending by area, keep first per GEOID — works across all pandas versions
    return (frame.sort_values(area_col, ascending=False)
                 .drop_duplicates(subset='GEOID', keep='first')[['GEOID','ZONECLASS']]
                 .reset_index(drop=True)
                 .rename(columns={'ZONECLASS': 'dominant'}))

# Drop any duplicate columns from repeated script runs
df_clean = df_clean.loc[:, ~df_clean.columns.duplicated()]

dom_old = get_dominant(df_clean, 'ShapeSTAre').rename(columns={'dominant': 'dominant_old'})
dom_new = get_dominant(df_clean, 'ShapeSTAre_corrected').rename(columns={'dominant': 'dominant_new'})

comp = dom_old.merge(dom_new, on='GEOID')
changed = comp[comp['dominant_old'] != comp['dominant_new']]

print(f"\nDominant zone changes after correction: "
      f"{len(changed)} / {len(comp)} block groups ({len(changed)/len(comp):.1%})")
print("\nSample of changed dominant zones:")
print(changed.head(10).to_string(index=False))

# ─────────────────────────────────────────────────────────────
# 7. IMPACT ON DOWNSTREAM ANALYSES
# ─────────────────────────────────────────────────────────────
print("\n── Impact summary ──")
print("Files that need to be rerun using tpazoning_clean.csv:")
print("  zoning_scraper.py            — reads ZON_CSV_PATH")
print("  embedding_pipeline.py        — reads ZON_CSV_PATH")
print("  tampa_active_travel_classifiers_v3/v4/v5/v6.py  — reads ZON_CSV_PATH")
print("  shapefile_maps_final.py      — reads zon_csv path")
print()
print("Update ZON_CSV_PATH / zon_csv path in each script to point to")
print("tpazoning_clean.csv and ensure all code uses 'ShapeSTAre_corrected'")
print("instead of 'ShapeSTAre' for area weighting.")
```
