```python
"""
Extract Tampa City Block Groups from Florida State Shapefile
=============================================================
Filters tl_2025_12_bg.shp (all Florida, 13,388 BGs) down to
City of Tampa only (367 BGs) by cross-referencing with the
GEOIDs present in the Tampa zoning parcel dataset.

Why this filter is correct:
  The zoning parcel data comes from the City of Tampa GeoHub and
  covers only parcels within the city boundary. Any census block
  group that contains at least one Tampa zoning parcel is by
  definition (at least partially) within the city. This gives
  367 block groups — the same set used throughout the analysis.

Required files (all in the same directory as this script):
  tl_2025_12_bg.shp    — Florida state block groups (full file)
  tl_2025_12_bg.dbf    — accompanying attribute table
  tl_2025_12_bg.shx    — shape index file
  tpazoning_clean.csv  — Tampa zoning parcels (for GEOID reference)
                         (or tpazoning.csv if clean version not yet available)

Outputs:
  tampa_bg.shp   — Tampa-only block groups (367 features)
  tampa_bg.dbf   — accompanying attribute table
  tampa_bg.shx   — shape index
  tampa_bg.prj   — projection file (copied from source)
"""

import os, struct, shutil
import pandas as pd

# ─────────────────────────────────────────────────────────────
# CONFIGURATION
# ─────────────────────────────────────────────────────────────
SHP_IN  = "raw_data/tl_2025_12_bg/tl_2025_12_bg.shp"
DBF_IN  = "raw_data/tl_2025_12_bg/tl_2025_12_bg.dbf"
SHX_IN  = 'raw_data/tl_2025_12_bg/tl_2025_12_bg.shx'
PRJ_IN  = 'raw_data/tl_2025_12_bg/tl_2025_12_bg.prj'

# Use clean file if available, otherwise fall back to original
ZON_CSV = 'tpazoning_clean.csv' if os.path.exists('tpazoning_clean.csv') \
          else 'tpazoning.csv'

SHP_OUT = 'tampa_bg.shp'
DBF_OUT = 'tampa_bg.dbf'
SHX_OUT = 'tampa_bg.shx'
PRJ_OUT = 'tampa_bg.prj'

# ─────────────────────────────────────────────────────────────
# 1. BUILD TAMPA GEOID SET
# ─────────────────────────────────────────────────────────────
print("Loading Tampa zoning GEOIDs...")
zon = pd.read_csv(ZON_CSV).dropna(subset=['ZONECLASS'])
tampa_geoids = set(zon['GEOID'].astype(str).unique())
print(f"  Tampa GEOIDs from zoning data: {len(tampa_geoids)}")

# ─────────────────────────────────────────────────────────────
# 2. READ DBF — identify which record indices to keep
# ─────────────────────────────────────────────────────────────
print("Reading DBF attribute table...")
with open(DBF_IN, 'rb') as f:
    header      = f.read(32)
    n_records   = struct.unpack('<I', header[4:8])[0]
    header_size = struct.unpack('<H', header[8:10])[0]
    record_size = struct.unpack('<H', header[10:12])[0]

    # Parse field descriptors
    fields = []
    while True:
        fd = f.read(32)
        if fd[0] == 0x0D or len(fd) < 32:
            break
        name  = fd[:11].replace(b'\x00', b'').decode('ascii', 'ignore').strip()
        ftype = chr(fd[11])
        flen  = fd[16]
        fields.append((name, ftype, flen))

    # Find GEOID field offset
    geoid_offset = 1  # deletion flag
    for name, ftype, flen in fields:
        if name == 'GEOID':
            break
        geoid_offset += flen
    geoid_len = next(flen for n, t, flen in fields if n == 'GEOID')

    # Identify which record indices are Tampa BGs
    f.seek(header_size)
    keep_indices = []
    for i in range(n_records):
        rec = f.read(record_size)
        if not rec or rec[0] == 0x1A:
            break
        geoid = rec[geoid_offset:geoid_offset + geoid_len].decode('ascii', 'ignore').strip()
        if geoid in tampa_geoids:
            keep_indices.append(i)

print(f"  Records to keep: {len(keep_indices)} / {n_records}")

# ─────────────────────────────────────────────────────────────
# 3. WRITE FILTERED DBF
# ─────────────────────────────────────────────────────────────
print("Writing tampa_bg.dbf...")
keep_set = set(keep_indices)
n_out    = len(keep_indices)

with open(DBF_IN, 'rb') as fin, open(DBF_OUT, 'wb') as fout:
    # Read full original header (includes field descriptors)
    orig_header = fin.read(header_size)

    # Patch record count in header (bytes 4-7, little-endian uint32)
    patched_header = bytearray(orig_header)
    struct.pack_into('<I', patched_header, 4, n_out)
    fout.write(bytes(patched_header))

    # Write only the kept records
    fin.seek(header_size)
    for i in range(n_records):
        rec = fin.read(record_size)
        if not rec:
            break
        if i in keep_set:
            fout.write(rec)

    # EOF marker
    fout.write(b'\x1a')

# ─────────────────────────────────────────────────────────────
# 4. READ SHX — get byte offset and content length per record
# ─────────────────────────────────────────────────────────────
print("Reading SHX index...")
with open(SHX_IN, 'rb') as f:
    shx_file_header = f.read(100)
    shx_records = []
    while True:
        rec = f.read(8)
        if len(rec) < 8:
            break
        offset      = struct.unpack('>I', rec[0:4])[0]  # in 16-bit words
        content_len = struct.unpack('>I', rec[4:8])[0]  # in 16-bit words
        shx_records.append((offset, content_len))

print(f"  SHX records: {len(shx_records)}")

# ─────────────────────────────────────────────────────────────
# 5. WRITE FILTERED SHP + SHX
# ─────────────────────────────────────────────────────────────
print("Writing tampa_bg.shp and tampa_bg.shx...")

with open(SHP_IN,  'rb') as shp_in, \
     open(SHP_OUT, 'wb') as shp_out, \
     open(SHX_OUT, 'wb') as shx_out:

    # Copy SHP file header (100 bytes), will patch file length later
    shp_file_header = shp_in.read(100)
    shp_out.write(shp_file_header)

    # Write placeholder SHX file header (will patch later)
    shx_out.write(shx_file_header)

    current_word_offset = 50  # 100 bytes = 50 16-bit words (file header)
    new_shx_records     = []

    for i in keep_indices:
        orig_offset, content_len = shx_records[i]

        # Seek to record in SHP and read it
        # SHP record = 8-byte header + content_len*2 bytes
        byte_offset = orig_offset * 2
        shp_in.seek(byte_offset)
        rec_header  = shp_in.read(8)
        rec_content = shp_in.read(content_len * 2)

        # Write to output SHP
        shp_out.write(rec_header)
        shp_out.write(rec_content)

        # Record new SHX entry
        new_shx_records.append((current_word_offset, content_len))
        current_word_offset += 4 + content_len  # 4 words = 8-byte header

    # Patch SHP file length (bytes 24-27, big-endian int32, in 16-bit words)
    total_shp_words = current_word_offset
    shp_out.seek(24)
    shp_out.write(struct.pack('>I', total_shp_words))

    # Write SHX records
    for offset, content_len in new_shx_records:
        shx_out.write(struct.pack('>I', offset))
        shx_out.write(struct.pack('>I', content_len))

    # Patch SHX file length (50 header words + 4 words per record)
    total_shx_words = 50 + 4 * len(new_shx_records)
    shx_out.seek(24)
    shx_out.write(struct.pack('>I', total_shx_words))

# ─────────────────────────────────────────────────────────────
# 6. COPY PRJ
# ─────────────────────────────────────────────────────────────
if os.path.exists(PRJ_IN):
    shutil.copy(PRJ_IN, PRJ_OUT)
    print("Copied: tampa_bg.prj")

# ─────────────────────────────────────────────────────────────
# 7. VERIFY
# ─────────────────────────────────────────────────────────────
print("\nVerification:")

# Count records in output DBF
with open(DBF_OUT, 'rb') as f:
    header   = f.read(32)
    n_out_dbf = struct.unpack('<I', header[4:8])[0]
print(f"  DBF records: {n_out_dbf}")

# Count records in output SHX
shx_size = os.path.getsize(SHX_OUT)
n_out_shx = (shx_size - 100) // 8
print(f"  SHX records: {n_out_shx}")

# Spot-check a few GEOIDs in the output DBF
with open(DBF_OUT, 'rb') as f:
    header      = f.read(32)
    hsize       = struct.unpack('<H', header[8:10])[0]
    rsize       = struct.unpack('<H', header[10:12])[0]
    f2 = open(DBF_IN, 'rb')
    f2.read(32)
    flds = []
    while True:
        fd = f2.read(32)
        if fd[0] == 0x0D or len(fd) < 32: break
        name  = fd[:11].replace(b'\x00',b'').decode('ascii','ignore').strip()
        ftype = chr(fd[11]); flen = fd[16]
        flds.append((name, ftype, flen))
    f2.close()
    f.seek(hsize)
    sample_geoids = []
    for _ in range(5):
        rec = f.read(rsize)
        if not rec: break
        pos = 1
        for name, ftype, flen in flds:
            val = rec[pos:pos+flen].decode('ascii','ignore').strip()
            if name == 'GEOID':
                sample_geoids.append(val)
            pos += flen
    print(f"  Sample GEOIDs in output: {sample_geoids}")
    all_in_tampa = all(g in tampa_geoids for g in sample_geoids)
    print(f"  All sample GEOIDs in Tampa set: {all_in_tampa}")

print(f"\n✓ Done — tampa_bg.shp contains {n_out_dbf} Tampa block groups")
print("  Use tampa_bg.shp in shapefile_maps_final.py by updating BG_SHP_PATH")
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Read shapefile
your = gpd.read_file('raw_data/shapefiles/Zoning/Zoning_District.shp')
tract = gpd.read_file("raw_data/tl_2025_12_bg/tl_2025_12_bg.shp")

# Coordinate system 1
tract = tract.to_crs(your.crs)

# Find the tract that intersects your shapefile
intersect = gpd.overlay(tract, your, how="intersection")

tract_plot = tract[tract["GEOID"].isin(intersect["GEOID"])]

# Generate label points
tract_plot["label_point"] = tract_plot.representative_point()

fig, ax = plt.subplots(figsize=(10,10))

# tract boundary
tract_plot.boundary.plot(ax=ax, color="blue", linewidth=0.8)

# your shapefile
your.boundary.plot(ax=ax, color="red", linewidth=1)

# Mark tract GEOID
for _, row in tract_plot.iterrows():
    ax.text(
row["label_point"].x,
row["label_point"].y,
row["GEOID"],
fontsize=8,
ha="center"
)

plt.show()
```
