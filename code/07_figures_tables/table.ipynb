```python
"""
Parcel-Level Output Table: Embeddings + Scraper Features + ACS
==============================================================
Produces a single CSV where each row is a zoning parcel from
tpazoning_clean.csv, enriched with:

  Zone-level columns (from the parcel's own ZONECLASS):
    - Raw 32-dim embedding (zone_emb_000 … zone_emb_031)
    - Scraper features (ped_score_norm, front_setback, lot_width, max_height)
    - Zone context label (Residential / Non-Residential / Other/PD)

  Block-group-level columns (from the parcel's GEOID):
    - Area-weighted 32-dim embedding (bg_emb_000 … bg_emb_031)
    - Area-weighted scraper features (ped_score_weighted, front_setback_avg, …)
    - Zoning entropy
    - ACS sociodemographics (median_hh_income, pct_bachelors_plus, …)

  Parcel-level geometry:
    - Area (ShapeSTAre_corrected)
    - Area share within BG (area_share)
    - Dominant zone flag (is_dominant)

Output:
  parcel_embedding_acs_table.csv

Run:
  python parcel_embedding_acs_table.py
"""

import numpy as np
import pandas as pd

# ─────────────────────────────────────────────────────────────
# PATHS
# ─────────────────────────────────────────────────────────────
ZON_CSV_PATH = 'processed_data/tpazoning_clean.csv'
FEAT_PATH    = 'zone_features_combined.csv'
EMB_PATH     = 'zone_embeddings.csv'
ACS_PATH     = 'acs_blockgroup_tampa.csv'

OUTPUT_PATH  = 'parcel_embedding_acs_table.csv'

# ─────────────────────────────────────────────────────────────
# ZONE CONTEXT DEFINITIONS
# ─────────────────────────────────────────────────────────────
RESIDENTIAL = [
    'RS-150','RS-100','RS-75','RS-60','RS-50',
    'RM-12','RM-16','RM-18','RM-24','RM-35','RM-50','RM-75',
    'RO','RO-1','SH-RS','SH-RS-A','SH-RM','SH-RO','SH-PD',
    'YC-2','YC-4','YC-8','YC-9',
]
NON_RESIDENTIAL = [
    'CG','CI','CN','OP','OP-1','IG','IH','CBD-1','CBD-2',
    'CD-1','CD-2','CD-3','NMU-35','NMU-24','NMU-16',
    'SH-CG','SH-CI','SH-CN','YC-1','YC-3','YC-5','YC-6','YC-7',
]

ACS_VARS = [
    'median_hh_income', 'total_pop', 'total_hh',
    'pct_no_vehicle', 'pct_bachelors_plus',
    'pct_white_nonhisp', 'pct_black_nonhisp', 'pct_hispanic',
    'pct_poverty', 'pct_owner_occ',
]

# ─────────────────────────────────────────────────────────────
# 1. LOAD DATA
# ─────────────────────────────────────────────────────────────
print('Loading data...')
zon_csv  = pd.read_csv(ZON_CSV_PATH).dropna(subset=['ZONECLASS'])
zone_feat = pd.read_csv(FEAT_PATH)
zone_emb = pd.read_csv(EMB_PATH)
acs      = pd.read_csv(ACS_PATH)

area_col = ('ShapeSTAre_corrected' if 'ShapeSTAre_corrected' in zon_csv.columns
            else 'ShapeSTAre')
emb_cols = [c for c in zone_emb.columns if c.startswith('emb_')]
n_dims   = len(emb_cols)

print(f'  Parcels:      {len(zon_csv):,}')
print(f'  Block groups: {zon_csv["GEOID"].nunique():,}')
print(f'  Zone classes:  {zon_csv["ZONECLASS"].nunique()}')
print(f'  Embedding dims: {n_dims}')

# ─────────────────────────────────────────────────────────────
# 2. ZONE-LEVEL JOINS (parcel's own ZONECLASS)
# ─────────────────────────────────────────────────────────────
print('\nJoining zone-level features to parcels...')

# Zone context
zon_csv['zone_context'] = 'Other/PD'
zon_csv.loc[zon_csv['ZONECLASS'].isin(RESIDENTIAL),     'zone_context'] = 'Residential'
zon_csv.loc[zon_csv['ZONECLASS'].isin(NON_RESIDENTIAL), 'zone_context'] = 'Non-Residential'

# Scraper features (zone-level)
scraper_cols = ['zone_class', 'ped_score_norm', 'front_setback',
                'lot_width', 'max_height', 'ped_score_total']
scraper_avail = [c for c in scraper_cols if c in zone_feat.columns]
zon_csv = zon_csv.merge(zone_feat[scraper_avail],
                        left_on='ZONECLASS', right_on='zone_class', how='left')
if 'zone_class' in zon_csv.columns:
    zon_csv.drop(columns=['zone_class'], inplace=True)

# Zone-level raw embedding
zone_emb_renamed = zone_emb[['zone_class'] + emb_cols].copy()
zone_emb_cols_renamed = {c: f'zone_{c}' for c in emb_cols}
zone_emb_renamed.rename(columns=zone_emb_cols_renamed, inplace=True)
zon_csv = zon_csv.merge(zone_emb_renamed,
                        left_on='ZONECLASS', right_on='zone_class', how='left')
if 'zone_class' in zon_csv.columns:
    zon_csv.drop(columns=['zone_class'], inplace=True)

n_matched = zon_csv[[f'zone_{emb_cols[0]}']].notna().sum().iloc[0]
print(f'  Zone embeddings matched: {n_matched:,} / {len(zon_csv):,} parcels')

# ─────────────────────────────────────────────────────────────
# 3. BLOCK-GROUP-LEVEL FEATURES
# ─────────────────────────────────────────────────────────────
print('\nComputing block-group-level features...')

# --- Area share and dominant zone flag ---
bg_totals = zon_csv.groupby('GEOID')[area_col].transform('sum')
zon_csv['area_share'] = zon_csv[area_col] / bg_totals

dominant_idx = zon_csv.groupby('GEOID')[area_col].idxmax()
zon_csv['is_dominant'] = False
zon_csv.loc[dominant_idx, 'is_dominant'] = True

# --- Area-weighted scraper features ---
def weighted_mean(group, val_col):
    mask = group[val_col].notna()
    if mask.sum() == 0:
        return np.nan
    return np.average(group.loc[mask, val_col],
                      weights=group.loc[mask, area_col])

scraper_agg = (zon_csv.groupby('GEOID')
               .apply(lambda g: pd.Series({
                   'bg_ped_score_weighted': weighted_mean(g, 'ped_score_norm'),
                   'bg_front_setback_avg':  weighted_mean(g, 'front_setback'),
                   'bg_lot_width_avg':      weighted_mean(g, 'lot_width'),
                   'bg_max_height_avg':     weighted_mean(g, 'max_height'),
               }), include_groups=False)
               .reset_index())

# --- Zoning entropy ---
def entropy_fn(group):
    total = group[area_col].sum()
    if total == 0:
        return 0.0
    p = group[area_col] / total
    raw = float(-(p[p > 0] * np.log(p[p > 0])).sum())
    return max(0.0, raw)

entropy = (zon_csv.groupby('GEOID')
           .apply(entropy_fn, include_groups=False)
           .reset_index()
           .rename(columns={0: 'bg_zoning_entropy'}))

# --- Area-weighted embeddings ---
emb_lookup = zone_emb.set_index('zone_class')[emb_cols].to_dict('index')
bg_emb_rows = []
for geoid, grp in zon_csv.groupby('GEOID'):
    total = grp[area_col].sum()
    vec   = np.zeros(n_dims)
    cov   = 0.0
    for _, row in grp.iterrows():
        z = row['ZONECLASS']
        if z in emb_lookup:
            w    = row[area_col] / total
            vec += np.array(list(emb_lookup[z].values())) * w
            cov += w
    if cov > 0:
        bg_emb_cols = {f'bg_{emb_cols[i]}': vec[i] for i in range(n_dims)}
        bg_emb_rows.append({'GEOID': geoid, 'bg_emb_coverage': round(cov, 4),
                            **bg_emb_cols})

bg_emb = pd.DataFrame(bg_emb_rows)
print(f'  BG embeddings computed: {len(bg_emb):,} block groups')

# --- Merge BG-level features ---
bg_features = (scraper_agg
               .merge(entropy, on='GEOID', how='outer')
               .merge(bg_emb,  on='GEOID', how='outer'))

# --- ACS ---
acs_avail = [c for c in ACS_VARS if c in acs.columns]
bg_features = bg_features.merge(acs[['GEOID'] + acs_avail],
                                on='GEOID', how='left')
print(f'  ACS variables joined: {len(acs_avail)}')

# ─────────────────────────────────────────────────────────────
# 4. FINAL MERGE: parcel ← BG features
# ─────────────────────────────────────────────────────────────
print('\nMerging BG features to parcels...')
out = zon_csv.merge(bg_features, on='GEOID', how='left')

# ─────────────────────────────────────────────────────────────
# 5. SELECT AND ORDER COLUMNS
# ─────────────────────────────────────────────────────────────
# Parcel identifiers
id_cols = ['OBJECTID', 'GEOID', 'ZONECLASS', 'zone_context', 'is_dominant',
           area_col, 'area_share']

# Zone-level features (parcel's own zone)
zone_scraper = ['ped_score_norm', 'front_setback', 'lot_width', 'max_height']
zone_scraper = [c for c in zone_scraper if c in out.columns]
zone_emb_out = [f'zone_{c}' for c in emb_cols]
zone_emb_out = [c for c in zone_emb_out if c in out.columns]

# BG-level features
bg_scraper = ['bg_ped_score_weighted', 'bg_front_setback_avg',
              'bg_lot_width_avg', 'bg_max_height_avg', 'bg_zoning_entropy']
bg_emb_out = [f'bg_{c}' for c in emb_cols]
bg_emb_out = [c for c in bg_emb_out if c in out.columns]
bg_meta    = ['bg_emb_coverage']

# ACS
acs_out = [c for c in acs_avail if c in out.columns]

col_order = id_cols + zone_scraper + zone_emb_out + bg_scraper + bg_meta + bg_emb_out + acs_out
col_order = [c for c in col_order if c in out.columns]
out = out[col_order]

# ─────────────────────────────────────────────────────────────
# 6. SAVE
# ─────────────────────────────────────────────────────────────
out.to_csv(OUTPUT_PATH, index=False)
print(f'\nSaved: {OUTPUT_PATH}')
print(f'  Rows: {len(out):,}')
print(f'  Columns: {len(out.columns)}')

# Column summary
print(f'\n  Column groups:')
print(f'    Parcel IDs & geometry:    {len(id_cols)}')
print(f'    Zone-level scraper:       {len(zone_scraper)}')
print(f'    Zone-level embedding:     {len(zone_emb_out)}')
print(f'    BG area-weighted scraper: {len(bg_scraper)}')
print(f'    BG area-weighted embedding: {len(bg_emb_out)} + coverage')
print(f'    ACS sociodemographics:    {len(acs_out)}')
print(f'    Total:                    {len(col_order)}')

# Quick sanity checks
print(f'\n  Sanity checks:')
print(f'    Zone contexts: {out["zone_context"].value_counts().to_dict()}')
print(f'    Dominant parcels: {out["is_dominant"].sum():,} '
      f'(should ≈ {out["GEOID"].nunique():,} unique BGs)')
print(f'    Embedding coverage: '
      f'mean={out["bg_emb_coverage"].mean():.3f}, '
      f'min={out["bg_emb_coverage"].min():.3f}')
print(f'    ACS nulls: {out[acs_out].isna().any(axis=1).sum():,} '
      f'parcels with any ACS missing')

print('\n✓ Done.')
```
