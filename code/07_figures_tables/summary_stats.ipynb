```python
#!/usr/bin/env python3
"""
summary_stats.py — Descriptive statistics (mean, std) for tampatrips_1.csv
==========================================================================
Usage:
    python summary_stats.py tampatrips_1.csv

Output:
    tampatrips_summary_stats.csv
    Console table of mean and std for each numeric variable
"""

import sys
import pandas as pd

# ── Load ────────────────────────────────────────────────────
fname = 'tampatrips_1.csv'  # <-- adjust path if needed
df = pd.read_csv(fname)

df.loc[df['education'] > 7, 'education'] = pd.NA

# Drop geometry and ID columns
drop_cols = [c for c in df.columns
             if c.lower().startswith(('origin_geom', 'dest_geom',
                                      'unnamed', 'personid', 'hh_id'))]
# Also drop block group FIPS (they're codes, not numeric measures)
drop_cols += [c for c in df.columns if c in ('o_bg', 'd_bg')]
df_num = df.drop(columns=[c for c in drop_cols if c in df.columns])

# Keep only numeric columns
df_num = df_num.select_dtypes(include='number')

# ── Compute stats ───────────────────────────────────────────
stats = pd.DataFrame({
    'n':      df_num.count(),
    'mean':   df_num.mean(),
    'std':    df_num.std(),
    'min':    df_num.min(),
    'median': df_num.median(),
    'max':    df_num.max(),
})
stats = stats.round(4)

print(f'\nSummary statistics for {fname}  (n = {len(df):,} trips)\n')
print(stats.to_string())

# ── Save ────────────────────────────────────────────────────
out = 'tampatrips_summary_stats.csv'
stats.to_csv(out)
print(f'\nSaved: {out}')
```
